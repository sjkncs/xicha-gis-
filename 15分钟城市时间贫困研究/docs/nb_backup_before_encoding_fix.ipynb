```python
# ============================================================================
# 莽聽聰莽漏露氓聦潞氓聼聼盲赂聨忙颅楼猫隆聦莽陆聭莽禄聹猫聨路氓聫聳
# ============================================================================

# 氓聧聴氓卤卤氓聦潞猫戮鹿莽聲聦忙隆聠茂录聢忙聣聥氓聤篓bbox茂录聦茅聙聜莽聰篓盲潞聨忙聴聽莽陆聭莽禄聹忙聢聳API茅聶聬氓聢露忙聝聟氓聠碌茂录聣
BBOX = {
    'north': 22.545,
    'south': 22.475,
    'east': 113.975,
    'west': 113.845
}

PLACE_NAME = "Nanshan District, Shenzhen, China"

# 盲录聵氓聟聢氓掳聺猫炉聲盲禄聨猫隆聦忙聰驴氓聦潞氓聬聧莽搂掳猫聨路氓聫聳
cache_path = os.path.join(BASE_DIR, 'osm_cache')
os.makedirs(cache_path, exist_ok=True)

print("忙颅拢氓聹篓盲禄聨 OpenStreetMap 猫聨路氓聫聳忙路卤氓聹鲁氓赂聜氓聧聴氓卤卤氓聦潞忙颅楼猫隆聦茅聛聯猫路炉莽陆聭莽禄聹...")
print("(茅娄聳忙卢隆猫驴聬猫隆聦茅聹聙猫娄聛盲赂聥猫陆陆忙聲掳忙聧庐茂录聦茅垄聞猫庐隆 30-60 莽搂聮)\n")

try:
    # 忙聳鹿忙鲁聲1: 忙聦聣猫隆聦忙聰驴氓聦潞氓聬聧莽搂掳猫聨路氓聫聳
    G_walk = ox.graph_from_place(
        PLACE_NAME,
        network_type='walk',
        simplify=True,
        clean_periphery=True
    )
    print(f"芒聹聯 忙聦聣猫隆聦忙聰驴氓聦潞猫聨路氓聫聳忙聢聬氓聤聼: {len(G_walk.nodes)} 猫聤聜莽聜鹿, {len(G_walk.edges)} 猫戮鹿")
    
except Exception as e:
    print(f"忙聳鹿忙鲁聲1氓陇卤猫麓楼: {e}")
    print("盲陆驴莽聰篓 BBOX 忙聳鹿氓录聫猫聨路氓聫聳...")
    
    try:
        # 忙聳鹿忙鲁聲2: 忙聦聣猫戮鹿莽聲聦忙隆聠猫聨路氓聫聳
        G_walk = ox.graph_from_bbox(
            bbox=(BBOX['north'], BBOX['south'], BBOX['east'], BBOX['west']),
            network_type='walk',
            simplify=True
        )
        print(f"芒聹聯 忙聦聣bbox猫聨路氓聫聳忙聢聬氓聤聼: {len(G_walk.nodes)} 猫聤聜莽聜鹿, {len(G_walk.edges)} 猫戮鹿")
        
    except Exception as e2:
        print(f"忙聳鹿忙鲁聲2氓陇卤猫麓楼: {e2}")
        print("盲陆驴莽聰篓氓陇職猫戮鹿氓陆垄忙聳鹿氓录聫猫聨路氓聫聳...")
        
        # 忙聳鹿忙鲁聲3: 猫聡陋氓庐職盲鹿聣氓陇職猫戮鹿氓陆垄茂录聢氓聧聴氓卤卤氓聦潞氓陇搂猫聡麓猫陆庐氓禄聯茂录聣
        from shapely.geometry import Polygon
        ns_polygon = Polygon([
            (113.845, 22.475), (113.975, 22.475),
            (113.975, 22.545), (113.845, 22.545)
        ])
        G_walk = ox.graph_from_polygon(
            ns_polygon,
            network_type='walk',
            simplify=True
        )
        print(f"芒聹聯 忙聦聣氓陇職猫戮鹿氓陆垄猫聨路氓聫聳忙聢聬氓聤聼: {len(G_walk.nodes)} 猫聤聜莽聜鹿, {len(G_walk.edges)} 猫戮鹿")
```

```python
# ============================================================================
# 猫路炉莽陆聭忙聲掳忙聧庐茅垄聞氓陇聞莽聬聠盲赂聨忙颅楼猫隆聦茅聙聼氓潞娄猫庐戮氓庐職
# ============================================================================

# 盲赂潞猫戮鹿忙路禄氓聤聽忙颅楼猫隆聦忙聴露茅聴麓氓卤聻忙聙搂茂录聢氓聢聠茅聮聼茂录聣
# 忙颅楼猫隆聦茅聙聼氓潞娄: 5.0 km/h = 83.33 m/min
WALK_SPEED_M_PER_MIN = 83.33

def add_travel_time(G, speed_mpm=WALK_SPEED_M_PER_MIN):
    """盲赂潞忙炉聫忙聺隆猫戮鹿忙路禄氓聤聽忙聴聟猫隆聦忙聴露茅聴麓(氓聢聠茅聮聼)氓卤聻忙聙搂"""
    for u, v, data in G.edges(data=True):
        if 'length' not in data:
            data['length'] = 0
        data['travel_time_min'] = data['length'] / speed_mpm
    return G

G_walk = add_travel_time(G_walk)

# 猫陆卢忙聧垄盲赂潞猫聤聜莽聜鹿GeoDataFrame氓聮聦猫戮鹿GeoDataFrame
nodes_gdf = ox.graph_to_gdfs(G_walk, edges=False, node_geometry=True)
edges_gdf = ox.graph_to_gdfs(G_walk, nodes=False, edge_geometry=False)

# 盲驴聺氓颅聵莽录聯氓颅聵
ox.save_graphml(G_walk, os.path.join(cache_path, 'nanshan_walk_network.graphml'))
ox.save_graphml(G_walk, os.path.join(cache_path, 'nanshan_walk_network.graphml.gz'))
print(f"芒聹聯 猫路炉莽陆聭氓路虏莽录聯氓颅聵: {cache_path}")

# 莽禄聼猫庐隆盲驴隆忙聛炉
print("\n" + "=" * 60)
print("猫路炉莽陆聭莽禄聼猫庐隆忙聭聵猫娄聛")
print("=" * 60)
print(f"猫聤聜莽聜鹿忙聲掳茅聡聫:   {len(G_walk.nodes):,}")
print(f"猫戮鹿忙聲掳茅聡聫:     {len(G_walk.edges):,}")
avg_edge_len = np.mean([d['length'] for u, v, d in G_walk.edges(data=True)])
total_len_km = sum([d['length'] for u, v, d in G_walk.edges(data=True)]) / 1000
print(f"氓鹿鲁氓聺聡猫戮鹿茅聲驴:   {avg_edge_len:.1f} m")
print(f"忙聙禄茅聛聯猫路炉茅聲驴氓潞娄: {total_len_km:.1f} km")
print(f"茅聛聯猫路炉莽卤禄氓聻聥:   {edges_gdf['highway'].value_counts().head(5).to_dict()}")
print(f"\n氓聺聬忙聽聡猫聦聝氓聸麓:")
print(f"  莽禄聫氓潞娄: [{nodes_gdf.geometry.x.min():.5f}, {nodes_gdf.geometry.x.max():.5f}]")
print(f"  莽潞卢氓潞娄: [{nodes_gdf.geometry.y.min():.5f}, {nodes_gdf.geometry.y.max():.5f}]")
```

```python
# ============================================================================
# 猫聥楼 OSM 忙聲掳忙聧庐盲赂聧猫露鲁茂录聦猫隆楼氓聟聟忙篓隆忙聥聼 POI茂录聢忙录聰莽陇潞莽聰篓茂录聣
# 氓庐聻茅聶聟氓聫聭猫隆篓忙聴露猫炉路忙聸驴忙聧垄盲赂潞茅芦聵氓戮路 API / 氓陇搂盲录聴莽聜鹿猫炉聞莽聹聼氓庐聻忙聲掳忙聧庐
# ============================================================================

def generate_supplementary_poi(bbox, n_per_type=30):
    """
    猫隆楼氓聟聟忙篓隆忙聥聼 POI茂录聢盲禄聟氓聹篓 OSM 忙聲掳忙聧庐莽篓聙莽聳聫忙聴露盲陆驴莽聰篓茂录聣
    忙篓隆忙聥聼氓聧聴氓卤卤氓聦潞莽聹聼氓庐聻 POI 莽漏潞茅聴麓氓聢聠氓赂聝莽聣鹿氓戮聛茂录職忙聽赂氓驴聝氓聦潞氓炉聠氓潞娄茅芦聵茫聙聛氓聭篓猫戮鹿氓炉聠氓潞娄盲陆聨
    """
    np.random.seed(42)
    
    # 忙篓隆忙聥聼氓聧聴氓卤卤氓聦潞 POI 氓炉聠氓潞娄盲赂颅氓驴聝茂录聢莽搂聭忙聤聙氓聸颅/氓聬聨忙碌路盲赂潞忙聽赂氓驴聝茂录聣
    center_lng, center_lat = 113.93, 22.53
    
    type_configs = [
        ('convenience', '盲戮驴氓聢漏氓潞聴', 40, ['08:00-23:00', '07:00-22:00', '24氓掳聫忙聴露']),
        ('pharmacy', '猫聧炉氓潞聴', 15, ['08:00-21:00', '09:00-22:00']),
        ('hospital', '氓聦禄茅聶垄', 3, ['08:00-18:00', '24氓掳聫忙聴露']),
        ('clinic', '猫炉聤忙聣聙', 8, ['08:00-20:00']),
        ('supermarket', '猫露聟氓赂聜', 12, ['08:00-22:00', '07:00-23:00']),
        ('bank', '茅聯露猫隆聦', 10, ['09:00-17:00', '09:00-17:30']),
        ('atm', 'ATM', 25, ['24氓掳聫忙聴露']),
        ('school', '氓颅娄忙聽隆', 8, ['07:30-17:30']),
        ('kindergarten', '氓鹿录氓聞驴氓聸颅', 10, ['07:00-18:00']),
        ('bus_stop', '氓聟卢盲潞陇莽芦聶', 80, ['24氓掳聫忙聴露']),
    ]
    
    records = []
    for ftype, fname, count, hours_list in type_configs:
        for i in range(count):
            # 盲陆驴莽聰篓茅芦聵忙聳炉氓聢聠氓赂聝忙篓隆忙聥聼莽漏潞茅聴麓猫聛職茅聸聠
            lng = np.random.normal(center_lng, 0.03)
            lat = np.random.normal(center_lat, 0.015)
            # 茅聶聬氓聢露氓聹篓 bbox 氓聠聟
            lng = np.clip(lng, bbox['west'], bbox['east'])
            lat = np.clip(lat, bbox['south'], bbox['north'])
            hours = np.random.choice(hours_list)
            is_24h = (hours == '24氓掳聫忙聴露')
            
            records.append({
                'name': f'{fname}_{i+1}',
                'facility_type': ftype,
                'lng': round(lng, 6),
                'lat': round(lat, 6),
                'business_hours': hours,
                'is_24h': is_24h,
                'source': 'simulated'
            })
    
    return pd.DataFrame(records)

# 氓聬聢氓鹿露 OSM 氓聮聦猫隆楼氓聟聟忙聲掳忙聧庐
if poi_osm is None or len(poi_osm) < 50:
    print("OSM 忙聲掳忙聧庐莽篓聙莽聳聫茂录聦猫隆楼氓聟聟忙篓隆忙聥聼忙聲掳忙聧庐...")
    poi_sup = generate_supplementary_poi(BBOX)
    poi_df = poi_sup
    print(f"猫隆楼氓聟聟忙篓隆忙聥聼忙聲掳忙聧庐: {len(poi_sup)} 盲赂陋")
else:
    # 氓炉鹿 OSM 忙聲掳忙聧庐忙路禄氓聤聽猫聬楼盲赂職忙聴露茅聴麓
    poi_osm['business_hours'] = 'unknown'
    poi_osm['is_24h'] = False
    poi_osm['source'] = 'OSM'
    poi_df = poi_osm.copy()

print(f"\n忙聹聙莽禄聢 POI 忙聲掳忙聧庐茅聸聠: {len(poi_df)} 盲赂陋猫庐戮忙聳陆")
print("\n猫庐戮忙聳陆莽卤禄氓聻聥氓聢聠氓赂聝:")
print(poi_df['facility_type'].value_counts().to_string())
```

**"猫聞聠氓录卤忙聙搂氓聫聽氓聤聽" (Vulnerability Stacking)**茂录職氓陆聯盲赂聙盲赂陋盲潞潞氓聬聦忙聴露氓卤聻盲潞聨氓陇職盲赂陋氓录卤氓聤驴莽戮陇盲陆聯忙聴露茂录聢氓娄聜氓聼聨盲赂颅忙聺聭莽職聞猫聙聛盲潞潞茂录聣茂录聦氓聟露氓聫炉猫戮戮忙聙搂莽潞娄忙聺聼氓聭聢忙聦聡忙聲掳莽潞搂忙聛露氓聦聳茂录聦猫聙聦茅聺聻莽潞驴忙聙搂氓聫聽氓聤聽茫聙聜

```python
# ============================================================================
# 氓录卤氓聤驴莽戮陇盲陆聯莽聰禄氓聝聫盲赂聨莽陇戮盲录職猫聞聠氓录卤忙聙搂猫炉聞氓聢聠莽鲁禄莽禄聼
# ============================================================================
"""
忙聹卢忙篓隆氓聺聴忙聻聞氓禄潞氓陇職莽禄麓猫聞聠氓录卤忙聙搂猫炉聞氓聢聠 (Multi-dimensional Vulnerability Index, MVI)茂录聦
氓掳聠忙炉聫盲赂陋氓掳聫氓聦潞/氓卤聟忙掳聭莽聜鹿莽職聞莽陇戮盲录職猫聞聠氓录卤忙聙搂茅聡聫氓聦聳盲赂潞氓聫炉忙聯聧盲陆聹莽職聞GIS忙聦聡忙聽聡茫聙聜

氓聫聜猫聙聝忙聳聡莽聦庐:
- Flanagan, B. E., et al. (2011). A social vulnerability index for disaster management.
  Journal of Homeland Security and Emergency Management.
- Cutter, S. L., et al. (2003). Social vulnerability to environmental hazards.
  Social Science Quarterly.
"""

class VulnerablePopulationProfiler:
    """
    氓陇職莽禄麓猫聞聠氓录卤忙聙搂猫炉聞氓聢聠氓聶篓
    
    莽禄麓氓潞娄忙聻聞忙聢聬:
    1. 盲陆聫忙聢驴猫聞聠氓录卤忙聙搂 (Housing Vulnerability)
    2. 莽陇戮盲录職莽禄聫忙碌聨猫聞聠氓录卤忙聙搂 (Socioeconomic Vulnerability)
    3. 莽漏潞茅聴麓氓聫炉猫戮戮猫聞聠氓录卤忙聙搂 (Spatial Access Vulnerability)
    4. 莽聰聼莽聬聠猫聞聠氓录卤忙聙搂 (Physiological Vulnerability)
    
    莽禄录氓聬聢猫聞聠氓录卤忙聙搂忙聦聡忙聲掳 MVI = w1*HV + w2*SEV + w3*SAV + w4*PV
    """
    
    def __init__(self):
        # 氓聬聞莽禄麓氓潞娄忙聺聝茅聡聧茂录聢氓聫炉猫掳聝忙聲麓茂录聦氓聫聧忙聵聽忙聰驴莽颅聳盲录聵氓聟聢莽潞搂茂录聣
        self.weights = {
            'housing': 0.25,        # 盲陆聫忙聢驴猫聞聠氓录卤忙聙搂
            'socioeconomic': 0.25,  # 莽陇戮盲录職莽禄聫忙碌聨猫聞聠氓录卤忙聙搂
            'spatial': 0.25,       # 莽漏潞茅聴麓氓聫炉猫戮戮猫聞聠氓录卤忙聙搂
            'physiological': 0.25    # 莽聰聼莽聬聠猫聞聠氓录卤忙聙搂
        }
    
    def compute_housing_vulnerability(self, community_type):
        """
        盲陆聫忙聢驴猫聞聠氓录卤忙聙搂 HV
        氓聼潞盲潞聨氓掳聫氓聦潞莽卤禄氓聻聥茂录聢氓聼聨盲赂颅忙聺聭茫聙聛盲驴聺茅職聹忙聢驴茫聙聛氓聲聠氓聯聛忙聢驴茫聙聛茅芦聵莽芦炉莽陇戮氓聦潞茂录聣莽職聞氓聨聠氓聫虏猫聞聠氓录卤忙聙搂猫炉聞盲录掳
        """
        housing_scores = {
            'urban_village': 0.85,     # 氓聼聨盲赂颅忙聺聭茂录職茅芦聵氓炉聠氓潞娄茫聙聛猫驴聺氓禄潞氓陇職茫聙聛忙露聢茅聵虏茅職聬忙聜拢
            'affordable_housing': 0.55, # 盲驴聺茅職聹忙聢驴茂录職盲陆聨忙聰露氓聟楼猫聛職茅聸聠盲陆聠猫庐戮忙聳陆猫戮聝忙聳掳
            'commodity_housing': 0.30,  # 氓聲聠氓聯聛忙聢驴茂录職盲赂颅莽颅聣猫聞聠氓录卤忙聙搂
            'high_end': 0.10            # 茅芦聵莽芦炉莽陇戮氓聦潞茂录職盲陆聨猫聞聠氓录卤忙聙搂
        }
        return housing_scores.get(community_type, 0.50)
    
    def compute_socioeconomic_vulnerability(self, community_type, population, built_year):
        """
        莽陇戮盲录職莽禄聫忙碌聨猫聞聠氓录卤忙聙搂 SEV
        
        忙聦聡忙聽聡茂录職
        - 盲潞潞氓聫拢氓炉聠氓潞娄茂录聢茅芦聵氓炉聠氓潞娄芒聠聮茅芦聵猫聞聠氓录卤忙聙搂茂录聣
        - 氓禄潞忙聢聬氓鹿麓盲禄拢茂录聢猫聙聛忙聴搂芒聠聮茅芦聵猫聞聠氓录卤忙聙搂茂录聣
        - 氓掳聫氓聦潞莽卤禄氓聻聥茂录聢氓聫聧忙聵聽氓卤聟忙掳聭莽陇戮盲录職莽禄聫忙碌聨氓聹掳盲陆聧茂录聣
        """
        # 盲潞潞氓聫拢氓炉聠氓潞娄氓陆聮盲赂聙氓聦聳茂录聢猫露聤氓陇搂猫露聤猫聞聠氓录卤茂录聣
        pop_density_score = np.clip(population / 5000, 0, 1)
        
        # 氓禄潞忙聢聬氓鹿麓盲禄拢茂录聢猫露聤猫聙聛猫露聤猫聞聠氓录卤茂录聣
        age_score = np.clip((2025 - built_year) / 40, 0, 1)
        
        # 莽卤禄氓聻聥忙聺聝茅聡聧
        type_weight = {
            'urban_village': 0.9,
            'affordable_housing': 0.7,
            'commodity_housing': 0.4,
            'high_end': 0.1
        }.get(community_type, 0.5)
        
        sev = 0.4 * type_weight + 0.3 * pop_density_score + 0.3 * age_score
        return np.clip(sev, 0, 1)
    
    def compute_spatial_vulnerability(self, dist_to_center, nearest_poi_dist):
        """
        莽漏潞茅聴麓氓聫炉猫戮戮猫聞聠氓录卤忙聙搂 SAV
        
        忙聦聡忙聽聡茂录職
        - 猫路聺氓聦潞盲赂颅氓驴聝猫路聺莽娄禄茂录聢猫露聤猫驴聹芒聠聮猫聞聠氓录卤忙聙搂猫露聤茅芦聵茂录聣
        - 忙聹聙猫驴聭猫庐戮忙聳陆猫路聺莽娄禄茂录聢猫露聤氓陇搂芒聠聮猫聞聠氓录卤忙聙搂猫露聤茅芦聵茂录聣
        """
        # 猫路聺盲赂颅氓驴聝猫路聺莽娄禄氓陆聮盲赂聙氓聦聳茂录聢氓聧聴氓卤卤氓聦潞氓聼聼莽潞娄10km脙聴10km茂录聦猫戮鹿莽聲聦莽潞娄7km茂录聣
        center_score = np.clip(dist_to_center / 5000, 0, 1)
        
        # 忙聹聙猫驴聭猫庐戮忙聳陆猫路聺莽娄禄茂录聢忙颅楼猫隆聦忙聴露茅聴麓忙聸驴盲禄拢茂录聣
        poi_score = np.clip(nearest_poi_dist / 1500, 0, 1)
        
        return 0.5 * center_score + 0.5 * poi_score
    
    def compute_physiological_vulnerability(self, community_type, population):
        """
        莽聰聼莽聬聠猫聞聠氓录卤忙聙搂 PV
        
        氓聼潞盲潞聨氓掳聫氓聦潞莽卤禄氓聻聥莽職聞氓鹿麓茅戮聞莽禄聯忙聻聞忙聨篓忙聳颅茂录聢忙篓隆忙聥聼茂录聣茂录職
        - 氓聼聨盲赂颅忙聺聭茂录職氓陇職盲赂潞氓陇聳忙聺楼氓聤隆氓路楼茅聺聮氓拢庐氓鹿麓茂录聦盲陆聠猫聙聛盲潞潞/氓聞驴莽芦楼忙聣聵莽庐隆茅聹聙忙卤聜茅芦聵
        - 盲驴聺茅職聹忙聢驴茂录職盲陆聨忙聰露氓聟楼氓庐露氓潞颅茂录聦猫聙聛盲潞潞氓聮聦氓聞驴莽芦楼忙炉聰盲戮聥猫戮聝茅芦聵
        - 茅芦聵莽芦炉莽陇戮氓聦潞茂录職盲赂颅茅聺聮氓鹿麓莽聶陆茅垄聠氓庐露氓潞颅
        """
        type_profiles = {
            'urban_village': {
                'elderly_ratio': 0.12,    # 12% 猫聙聛氓鹿麓盲潞潞茂录聢茅職聫猫驴聛猫聙聛盲潞潞茂录聣
                'child_ratio': 0.18,     # 18% 氓聞驴莽芦楼茂录聢莽聲聶氓庐聢氓聞驴莽芦楼茂录聣
                'disability_ratio': 0.035 # 3.5% 忙庐聥茅職聹盲潞潞氓拢芦
            },
            'affordable_housing': {
                'elderly_ratio': 0.22,    # 盲驴聺茅職聹忙聢驴猫聙聛盲潞潞忙炉聰盲戮聥忙聸麓茅芦聵
                'child_ratio': 0.20,
                'disability_ratio': 0.042
            },
            'commodity_housing': {
                'elderly_ratio': 0.15,
                'child_ratio': 0.12,
                'disability_ratio': 0.028
            },
            'high_end': {
                'elderly_ratio': 0.08,
                'child_ratio': 0.10,
                'disability_ratio': 0.020
            }
        }
        
        profile = type_profiles.get(community_type, type_profiles['commodity_housing'])
        
        # 莽聰聼莽聬聠猫聞聠氓录卤忙聙搂 = 氓聤聽忙聺聝猫聞聠氓录卤莽戮陇盲陆聯忙炉聰盲戮聥
        # 猫聙聛氓鹿麓盲潞潞忙聺聝茅聡聧1.0茂录聦忙庐聥茅職聹盲潞潞氓拢芦忙聺聝茅聡聧1.2茂录聢氓庐聦氓聟篓盲戮聺猫碌聳猫庐戮忙聳陆茂录聣茂录聦氓聞驴莽芦楼忙聺聝茅聡聧0.8
        pv = (1.0 * profile['elderly_ratio'] + 
              1.2 * profile['disability_ratio'] + 
              0.8 * profile['child_ratio'])
        
        return np.clip(pv / 0.5, 0, 1)  # 氓陆聮盲赂聙氓聦聳
    
    def profile_community(self, row, dist_to_center=None, nearest_poi_dist=None):
        """
        氓炉鹿氓聧聲盲赂陋氓掳聫氓聦潞猫驴聸猫隆聦氓庐聦忙聲麓猫聞聠氓录卤忙聙搂莽聰禄氓聝聫
        
        猫驴聰氓聸聻: dict茂录聦氓聦聟氓聬芦氓聬聞莽禄麓氓潞娄猫炉聞氓聢聠氓聮聦莽禄录氓聬聢MVI
        """
        ctype = row.get('community_type', 'commodity_housing')
        pop = row.get('population', 2000)
        year = row.get('built_year', 2010)
        
        hv = self.compute_housing_vulnerability(ctype)
        sev = self.compute_socioeconomic_vulnerability(ctype, pop, year)
        pv = self.compute_physiological_vulnerability(ctype, pop)
        
        if dist_to_center is not None and nearest_poi_dist is not None:
            sav = self.compute_spatial_vulnerability(dist_to_center, nearest_poi_dist)
        else:
            # 盲陆驴莽聰篓茅禄聵猫庐陇氓聙录茂录聢氓聬聨莽禄颅氓聫炉莽聰篓氓庐聻茅聶聟猫庐隆莽庐聴忙聸驴盲禄拢茂录聣
            sav = 0.50
        
        # 莽禄录氓聬聢猫聞聠氓录卤忙聙搂忙聦聡忙聲掳 MVI
        mvi = (self.weights['housing'] * hv +
               self.weights['socioeconomic'] * sev +
               self.weights['spatial'] * sav +
               self.weights['physiological'] * pv)
        
        return {
            'HV': hv,
            'SEV': sev,
            'SAV': sav,
            'PV': pv,
            'MVI': mvi
        }
    
    def profile_all_communities(self, gdf, dist_to_center_dict=None, nearest_poi_dict=None):
        """
        氓炉鹿忙聣聙忙聹聣氓掳聫氓聦潞忙聣鹿茅聡聫猫庐隆莽庐聴猫聞聠氓录卤忙聙搂莽聰禄氓聝聫
        """
        results = []
        for _, row in gdf.iterrows():
            dist_c = dist_to_center_dict.get(row['community_id']) if dist_to_center_dict else None
            poi_d = nearest_poi_dict.get(row['community_id']) if nearest_poi_dict else None
            profile = self.profile_community(row, dist_c, poi_d)
            results.append(profile)
        
        profile_df = pd.DataFrame(results)
        result_gdf = gdf.copy()
        for col in profile_df.columns:
            result_gdf[col] = profile_df[col].values
        
        return result_gdf
    
    def get_group_statistics(self, gdf, group_col='community_type'):
        """
        忙聦聣氓掳聫氓聦潞莽卤禄氓聻聥猫戮聯氓聡潞猫聞聠氓录卤忙聙搂莽禄聼猫庐隆忙聭聵猫娄聛
        """
        groups = {
            'urban_village': '氓聼聨盲赂颅忙聺聭',
            'affordable_housing': '盲驴聺茅職聹忙聢驴',
            'commodity_housing': '氓聲聠氓聯聛忙聢驴',
            'high_end': '茅芦聵莽芦炉莽陇戮氓聦潞'
        }
        
        summary = []
        for gtype, glabel in groups.items():
            subset = gdf[gdf[group_col] == gtype]
            if len(subset) == 0:
                continue
            summary.append({
                '莽卤禄氓聻聥': glabel,
                '氓掳聫氓聦潞忙聲掳': len(subset),
                'HV_氓聺聡氓聙录': subset['HV'].mean(),
                'SEV_氓聺聡氓聙录': subset['SEV'].mean(),
                'SAV_氓聺聡氓聙录': subset['SAV'].mean(),
                'PV_氓聺聡氓聙录': subset['PV'].mean(),
                'MVI_氓聺聡氓聙录': subset['MVI'].mean(),
                'MVI_盲赂颅盲陆聧忙聲掳': subset['MVI'].median(),
                'MVI_忙聹聙氓陇搂氓聙录': subset['MVI'].max()
            })
        
        return pd.DataFrame(summary).sort_values('MVI_氓聺聡氓聙录', ascending=False)


# 忙聣搂猫隆聦猫聞聠氓录卤忙聙搂猫炉聞氓聢聠
profiler = VulnerablePopulationProfiler()
communities_gdf = profiler.profile_all_communities(communities_gdf)

print("=" * 70)
print("氓录卤氓聤驴莽戮陇盲陆聯猫聞聠氓录卤忙聙搂氓聢聠忙聻聬 芒聙聰 氓陇職莽禄麓猫聞聠氓录卤忙聙搂忙聦聡忙聲掳 (MVI)")
print("=" * 70)

mvi_stats = profiler.get_group_statistics(communities_gdf)
print("\n忙聦聣氓掳聫氓聦潞莽卤禄氓聻聥莽職聞猫聞聠氓录卤忙聙搂氓聺聡氓聙录忙聨聮氓潞聫:")
print(mvi_stats.to_string(index=False))

print("\n" + "-" * 70)
print("氓聟鲁茅聰庐氓聫聭莽聨掳茂录職猫掳聛忙聵炉忙聹聙猫聞聠氓录卤莽職聞莽戮陇盲陆聯茂录聼")
print("-" * 70)

high_vul = communities_gdf.nlargest(10, 'MVI')[['name', 'community_type', 'MVI', 'HV', 'SEV', 'PV']]
print("\n猫聞聠氓录卤忙聙搂忙聹聙茅芦聵莽職聞10盲赂陋氓掳聫氓聦潞:")
cn_map = {'urban_village':'氓聼聨盲赂颅忙聺聭','affordable_housing':'盲驴聺茅職聹忙聢驴',
          'commodity_housing':'氓聲聠氓聯聛忙聢驴','high_end':'茅芦聵莽芦炉莽陇戮氓聦潞'}
high_vul['莽卤禄氓聻聥'] = high_vul['community_type'].map(cn_map)
print(high_vul[['name','莽卤禄氓聻聥','MVI','HV','SEV','PV']].rename(
    columns={'MVI':'莽禄录氓聬聢猫聞聠氓录卤忙聙搂','HV':'盲陆聫忙聢驴','SEV':'莽陇戮盲录職莽禄聫忙碌聨','PV':'莽聰聼莽聬聠'}
).to_string(index=False))

print("\n" + "-" * 70)
print("猫聞聠氓录卤忙聙搂氓聫聽氓聤聽氓聢聠忙聻聬 芒聙聰 氓陇職茅聡聧猫聞聠氓录卤忙聙搂猫炉聠氓聢芦")
print("-" * 70)
communities_gdf['is_elderly_dominated'] = communities_gdf['community_type'].isin(
    ['urban_village', 'affordable_housing'])
communities_gdf['has_high_physiological'] = communities_gdf['PV'] > communities_gdf['PV'].quantile(0.75)
communities_gdf['has_high_housing'] = communities_gdf['HV'] > communities_gdf['HV'].quantile(0.75)
communities_gdf['is_vulnerability_stacked'] = (
    communities_gdf['is_elderly_dominated'] & 
    communities_gdf['has_high_physiological']
)
stacked_count = communities_gdf['is_vulnerability_stacked'].sum()
print(f"猫聞聠氓录卤忙聙搂氓聫聽氓聤聽氓掳聫氓聦潞茂录聢氓聼聨盲赂颅忙聺聭/盲驴聺茅職聹忙聢驴 + 茅芦聵莽聰聼莽聬聠猫聞聠氓录卤忙聙搂茂录聣: {stacked_count} 盲赂陋 ({stacked_count/len(communities_gdf)*100:.1f}%)")
print("\n猫驴聶盲潞聸氓掳聫氓聦潞忙聵炉忙聰驴莽颅聳氓鹿虏茅垄聞莽職聞忙聹聙盲录聵氓聟聢莽聸庐忙聽聡莽戮陇盲陆聯茫聙聜")
```

盲赂陇忙颅楼莽搂禄氓聤篓忙聬聹莽麓垄忙鲁聲 (Two-Step Floating Catchment Area, 2SFCA) 莽聰卤 Luo & Wang (2003) 忙聫聬氓聡潞茂录聦
忙聵炉氓聼聨氓赂聜猫庐戮忙聳陆氓聫炉猫戮戮忙聙搂莽聽聰莽漏露盲赂颅猫垄芦氓录聲莽聰篓忙聹聙氓鹿驴忙鲁聸莽職聞忙篓隆氓聻聥盲鹿聥盲赂聙 (*Transportation Research Part D*, *Health & Place* 莽颅聣茅隆露氓聢聤)茫聙聜

**莽卢卢盲赂聙忙颅楼**茂录職氓炉鹿忙炉聫盲赂陋猫庐戮忙聳陆 $j$茂录聦猫庐隆莽庐聴氓聟露忙聹聧氓聤隆猫聝陆氓聤聸 $R_j$茂录職

$$R_j = \frac{S_j}{\sum_{k \in \{d_{kj} \leq d_0\}} P_k}$$

氓聟露盲赂颅 $S_j$ 盲赂潞猫庐戮忙聳陆 $j$ 莽職聞盲戮聸莽禄聶猫搂聞忙篓隆茂录聢氓聫炉莽聰篓茅聺垄莽搂炉/氓潞聤盲陆聧忙聲掳/猫炉聞氓聢聠莽颅聣茂录聣茂录聦
$P_k$ 盲赂潞忙聬聹莽麓垄氓聧聤氓戮聞 $d_0$ 氓聠聟氓卤聟盲陆聫氓聧聲氓聟聝 $k$ 莽職聞盲潞潞氓聫拢茂录聦
$d_{kj}$ 盲赂潞 $k$ 氓聢掳 $j$ 莽職聞猫路炉莽陆聭猫路聺莽娄禄茫聙聜

**莽卢卢盲潞聦忙颅楼**茂录職氓炉鹿忙炉聫盲赂陋氓卤聟盲陆聫氓聧聲氓聟聝 $i$茂录聦猫庐隆莽庐聴氓聟露氓聫炉猫戮戮忙聙搂 $A_i$茂录職

$$A_i = \sum_{j \in \{d_{ij} \leq d_0\}} R_j$$

氓聬芦盲鹿聣茂录職氓聹篓 $i$ 莽職聞忙颅楼猫隆聦忙聬聹莽麓垄氓聧聤氓戮聞 $d_0$ 氓聠聟莽職聞忙聣聙忙聹聣猫庐戮忙聳陆莽職聞忙聹聧氓聤隆猫聝陆氓聤聸盲鹿聥氓聮聦茫聙聜

- 茅聹聙忙卤聜忙聦聡忙聽聡 $P_k$茂录職氓掳聫氓聦潞盲潞潞氓聫拢

- 猫路聺莽娄禄猫隆掳氓聡聫茂录職盲陆驴莽聰篓盲潞聦氓聙录猫隆掳氓聡聫茂录聢氓聧聤氓戮聞氓聠聟=1茂录聦氓聧聤氓戮聞氓陇聳=0茂录聣

```python
# ============================================================================
# 猫路炉莽陆聭猫路聺莽娄禄猫庐隆莽庐聴氓路楼氓聟路
# ============================================================================

class NetworkDistanceCalculator:
    """
    氓聼潞盲潞聨 OSMnx 猫路炉莽陆聭莽職聞忙聹聙莽聼颅猫路炉氓戮聞猫路聺莽娄禄猫庐隆莽庐聴氓聶篓
    盲陆驴莽聰篓 Dijkstra 莽庐聴忙鲁聲茂录聦忙聰炉忙聦聛氓陇搂猫搂聞忙篓隆 OD 莽聼漏茅聵碌莽職聞忙聣鹿茅聡聫猫庐隆莽庐聴
    """
    
    def __init__(self, G, walk_speed_mpm=WALK_SPEED_M_PER_MIN):
        self.G = G
        self.walk_speed = walk_speed_mpm
        self._build_node_tree()
        self._node_to_graphidx = {node: i for i, node in enumerate(G.nodes)}
        self._graphidx_to_node = {i: node for i, node in enumerate(G.nodes)}
        self._precompute_neighbors()
        
    def _build_node_tree(self):
        """忙聻聞氓禄潞莽芦聥氓聧鲁忙聼楼猫炉垄忙聹聙猫驴聭猫路炉莽陆聭猫聤聜莽聜鹿莽職聞 kd-tree"""
        node_list = list(self.G.nodes)
        coords = np.array([(self.G.nodes[n]['x'], self.G.nodes[n]['y']) for n in node_list])
        self.node_tree = cKDTree(coords)
        self.node_list = node_list
        self.node_coords = coords
        
    def _precompute_neighbors(self):
        """盲赂潞猫路聺莽娄禄茅聵聢氓聙录猫庐隆莽庐聴茅垄聞氓颅聵茅聜禄氓卤聟猫聤聜莽聜鹿茂录聢氓聤聽茅聙聼忙聣鹿茅聡聫忙聼楼猫炉垄茂录聣"""
        pass  # 忙聦聣茅聹聙氓庐聻莽聨掳
        
    def find_nearest_node(self, lng, lat):
        """忙聣戮氓聢掳猫路聺莽娄禄盲禄禄忙聞聫氓聺聬忙聽聡忙聹聙猫驴聭莽職聞 OSM 猫聤聜莽聜鹿 ID"""
        dist, idx = self.node_tree.query([lng, lat])
        return self.node_list[idx], dist, idx
        
    def network_distance_m(self, origin_lng, origin_lat, dest_lng, dest_lat):
        """
        猫庐隆莽庐聴盲赂陇莽聜鹿茅聴麓莽職聞忙聹聙莽聼颅猫路炉莽陆聭猫路聺莽娄禄茂录聢莽卤鲁茂录聣
        
        氓聫聜忙聲掳:
            origin_lng, origin_lat: 猫碌路莽聜鹿氓聺聬忙聽聡
            dest_lng, dest_lat: 莽禄聢莽聜鹿氓聺聬忙聽聡
        猫驴聰氓聸聻:
            float: 猫路炉莽陆聭猫路聺莽娄禄茂录聢莽卤鲁茂录聣茂录聦猫聥楼盲赂聧氓聫炉猫戮戮氓聢聶猫驴聰氓聸聻 np.inf
        """
        orig_node, _, _ = self.find_nearest_node(origin_lng, origin_lat)
        dest_node, _, _ = self.find_nearest_node(dest_lng, dest_lat)
        
        if orig_node == dest_node:
            return 0.0
        
        try:
            # 猫庐隆莽庐聴忙聹聙莽聼颅猫路炉氓戮聞茅聲驴氓潞娄茂录聢忙聦聣猫戮鹿茅聲驴氓潞娄氓聤聽忙聺聝茂录聣
            length = nx.shortest_path_length(
                self.G, orig_node, dest_node, weight='length'
            )
            return float(length)
        except (nx.NetworkXNoPath, nx.NodeNotFound):
            # 氓娄聜忙聻聹猫路炉莽陆聭盲赂聧猫驴聻茅聙職茂录聦盲录掳莽庐聴忙卢搂忙掳聫猫路聺莽娄禄 * 1.4茂录聢氓聟赂氓聻聥莽禄聲猫隆聦莽鲁禄忙聲掳茂录聣
            euclidean_dist = self._haversine_m(origin_lng, origin_lat, dest_lng, dest_lat)
            return euclidean_dist * 1.4
            
    def network_travel_time(self, origin_lng, origin_lat, dest_lng, dest_lat):
        """猫庐隆莽庐聴盲赂陇莽聜鹿茅聴麓莽職聞忙颅楼猫隆聦忙聴露茅聴麓茂录聢氓聢聠茅聮聼茂录聣"""
        dist = self.network_distance_m(origin_lng, origin_lat, dest_lng, dest_lat)
        return dist / self.walk_speed
        
    def _haversine_m(self, lng1, lat1, lng2, lat2):
        """Haversine 氓聟卢氓录聫猫庐隆莽庐聴莽聬聝茅聺垄猫路聺莽娄禄茂录聢莽卤鲁茂录聣"""
        R = 6371000
        phi1, phi2 = math.radians(lat1), math.radians(lat2)
        dphi = math.radians(lat2 - lat1)
        dlambda = math.radians(lng2 - lng1)
        a = math.sin(dphi/2)**2 + math.cos(phi1)*math.cos(phi2)*math.sin(dlambda/2)**2
        return 2 * R * math.atan2(math.sqrt(a), math.sqrt(1-a))
        
    def build_od_matrix(self, origins_df, destinations_df, 
                        lng_col='lng', lat_col='lat',
                        batch_size=500, verbose=True):
        """
        忙聣鹿茅聡聫猫庐隆莽庐聴 OD 忙聢聬忙聹卢莽聼漏茅聵碌
        盲陆驴莽聰篓 Dijkstra 忙聹聙莽聼颅猫路炉茂录聦猫庐隆莽庐聴忙聣聙忙聹聣 origin-destination 氓炉鹿莽職聞猫路聺莽娄禄
        
        氓聫聜忙聲掳:
            origins_df: 猫碌路莽聜鹿 DataFrame
            destinations_df: 莽禄聢莽聜鹿 DataFrame
            batch_size: 忙炉聫忙聣鹿猫庐隆莽庐聴莽職聞猫庐掳氓陆聲忙聲掳
        猫驴聰氓聸聻:
            np.ndarray: shape (n_origins, n_destinations)茂录聦氓聧聲盲陆聧茂录職莽卤鲁
        """
        n_o = len(origins_df)
        n_d = len(destinations_df)
        od_matrix = np.full((n_o, n_d), np.inf)
        
        print(f"忙聻聞氓禄潞 OD 莽聼漏茅聵碌: {n_o} 脙聴 {n_d} = {n_o*n_d:,} 氓炉鹿茂录聢茅垄聞猫庐隆 {n_o*n_d/1000:.0f}s茂录聣")
        
        start = time.time()
        for i, (_, orig) in enumerate(origins_df.iterrows()):
            orig_node, _, _ = self.find_nearest_node(orig[lng_col], orig[lat_col])
            
            for j, (_, dest) in enumerate(destinations_df.iterrows()):
                dest_node, _, _ = self.find_nearest_node(dest[lng_col], dest[lat_col])
                
                if orig_node == dest_node:
                    od_matrix[i, j] = 0.0
                    continue
                
                try:
                    od_matrix[i, j] = nx.shortest_path_length(
                        self.G, orig_node, dest_node, weight='length'
                    )
                except (nx.NetworkXNoPath, nx.NodeNotFound):
                    eucl = self._haversine_m(
                        orig[lng_col], orig[lat_col], dest[lng_col], dest[lat_col]
                    )
                    od_matrix[i, j] = eucl * 1.4
            
            if (i + 1) % 10 == 0 and verbose:
                elapsed = time.time() - start
                rate = (i+1) / elapsed
                eta = (n_o - i - 1) / rate
                print(f"  猫驴聸氓潞娄: {i+1}/{n_o} ({100*(i+1)/n_o:.1f}%)  ETA: {eta:.0f}s")
        
        print(f"  芒聹聯 OD 莽聼漏茅聵碌忙聻聞氓禄潞氓庐聦忙聢聬茂录聦猫聙聴忙聴露 {time.time()-start:.1f}s")
        return od_matrix
        
    def build_od_matrix_vectorized(self, origins_df, destinations_df,
                                   lng_col='lng', lat_col='lat'):
        """
        氓聬聭茅聡聫氓聦聳 OD 莽聼漏茅聵碌忙聻聞氓禄潞茂录聢盲录聵氓聟聢盲陆驴莽聰篓茂录聦茅聙聬氓炉鹿猫庐隆莽庐聴 Dijkstra 猫路聺莽娄禄茂录聣
        茅聙聜莽聰篓盲潞聨盲赂颅莽颅聣猫搂聞忙篓隆忙聲掳忙聧庐茅聸聠茂录聢芒聣陇200脙聴200茂录聣
        """
        n_o, n_d = len(origins_df), len(destinations_df)
        print(f"忙聻聞氓禄潞氓聬聭茅聡聫氓聦聳 OD 莽聼漏茅聵碌: {n_o} 脙聴 {n_d} ...")
        
        # 茅垄聞猫庐隆莽庐聴忙聣聙忙聹聣猫碌路莽聜鹿莽職聞忙聹聙猫驴聭猫聤聜莽聜鹿
        origin_nodes = []
        origin_tree = cKDTree(self.node_coords)
        for _, row in origins_df.iterrows():
            node, _, _ = self.find_nearest_node(row[lng_col], row[lat_col])
            origin_nodes.append(node)
        
        # 茅垄聞猫庐隆莽庐聴忙聣聙忙聹聣莽禄聢莽聜鹿莽職聞忙聹聙猫驴聭猫聤聜莽聜鹿
        dest_nodes = []
        for _, row in destinations_df.iterrows():
            node, _, _ = self.find_nearest_node(row[lng_col], row[lat_col])
            dest_nodes.append(node)
        
        od_matrix = np.zeros((n_o, n_d))
        
        for i, orig_node in enumerate(origin_nodes):
            try:
                lengths = nx.single_source_dijkstra_path_length(
                    self.G, orig_node, weight='length'
                )
                for j, dest_node in enumerate(dest_nodes):
                    od_matrix[i, j] = lengths.get(dest_node, np.inf)
            except Exception:
                od_matrix[i, :] = np.inf
        
        # 氓陇聞莽聬聠忙聴聽猫路炉氓戮聞忙聝聟氓聠碌
        mask = od_matrix == 0
        for i in range(n_o):
            for j in range(n_d):
                if mask[i, j] and i != j:
                    orig = origins_df.iloc[i]
                    dest = destinations_df.iloc[j]
                    eucl = self._haversine_m(orig[lng_col], orig[lat_col], dest[lng_col], dest[lat_col])
                    od_matrix[i, j] = eucl * 1.4
        
        print(f"  芒聹聯 氓聬聭茅聡聫氓聦聳 OD 莽聼漏茅聵碌氓庐聦忙聢聬: {np.sum(np.isfinite(od_matrix))}/{n_o*n_d} 忙聹聣忙聲聢氓炉鹿")
        return od_matrix

# 氓聢聺氓搂聥氓聦聳猫路炉莽陆聭猫路聺莽娄禄猫庐隆莽庐聴氓聶篓
dist_calc = NetworkDistanceCalculator(G_walk)
print("猫路炉莽陆聭猫路聺莽娄禄猫庐隆莽庐聴氓聶篓氓聢聺氓搂聥氓聦聳氓庐聦忙聢聬")

# 氓驴芦茅聙聼忙碌聥猫炉聲
test_dist = dist_calc.network_distance_m(113.93, 22.53, 113.95, 22.54)
test_time = dist_calc.network_travel_time(113.93, 22.53, 113.95, 22.54)
print(f"忙碌聥猫炉聲猫路炉氓戮聞: 猫路聺莽娄禄={test_dist:.0f}m, 忙颅楼猫隆聦忙聴露茅聴麓={test_time:.1f}min")
```

```python
# ============================================================================
# 2SFCA 氓聫炉猫戮戮忙聙搂猫庐隆莽庐聴氓录聲忙聯聨
# ============================================================================

class TwoStepFloatingCatchmentArea:
    """
    盲赂陇忙颅楼莽搂禄氓聤篓忙聬聹莽麓垄忙鲁聲 (2SFCA) 氓庐聻莽聨掳
    
    氓聫聜猫聙聝忙聳聡莽聦庐:
    - Luo, W., & Wang, F. (2003). Measures of spatial accessibility to health care 
      in a GIS environment. International Journal of Geofraphy Information Science.
    - Luo, W., & Qi, Y. (2009). An enhanced two-step floating catchment area 
      method for analyzing spatial access to health care services. Papers in Regional Science.
    """
    
    def __init__(self, search_radius_m=1250, 
                 supply_col='supply', demand_col='population',
                 dist_matrix=None):
        """
        氓聫聜忙聲掳:
            search_radius_m: 忙聬聹莽麓垄氓聧聤氓戮聞茂录聢莽卤鲁茂录聣茂录聦茅禄聵猫庐陇 1250m茂录聢忙颅楼猫隆聦15氓聢聠茅聮聼 @ 5km/h茂录聣
            supply_col: 猫庐戮忙聳陆盲戮聸莽禄聶忙聦聡忙聽聡氓聢聴氓聬聧
            demand_col: 氓掳聫氓聦潞茅聹聙忙卤聜忙聦聡忙聽聡氓聢聴氓聬聧
            dist_matrix: 茅垄聞猫庐隆莽庐聴莽職聞 OD 猫路聺莽娄禄莽聼漏茅聵碌茂录聢numpy 2D array茂录聣
        """
        self.search_radius = search_radius_m
        self.supply_col = supply_col
        self.demand_col = demand_col
        self.od_matrix = dist_matrix
        
    def step1_supply_ratio(self, facilities_df, od_matrix):
        """
        莽卢卢盲赂聙忙颅楼茂录職猫庐隆莽庐聴忙炉聫盲赂陋猫庐戮忙聳陆莽職聞忙聹聧氓聤隆猫聝陆氓聤聸忙炉聰 R_j
        
        R_j = S_j / sum(P_k) for all k where d_kj <= d_0
        
        氓聧鲁茂录職猫庐戮忙聳陆盲戮聸莽禄聶 / 猫聬陆氓聹篓氓聟露忙聹聧氓聤隆氓聧聤氓戮聞氓聠聟莽職聞忙聙禄盲潞潞氓聫拢
        """
        n_fac = len(facilities_df)
        R_j = np.zeros(n_fac)
        
        for j in range(n_fac):
            # 忙聣戮氓聢掳忙聣聙忙聹聣猫聬陆氓聹篓猫庐戮忙聳陆 j 忙聹聧氓聤隆氓聧聤氓戮聞氓聠聟莽職聞氓掳聫氓聦潞
            reachable_mask = (od_matrix[:, j] <= self.search_radius) & np.isfinite(od_matrix[:, j])
            total_demand = facilities_df.iloc[reachable_mask][self.demand_col].sum()
            if total_demand > 0:
                R_j[j] = facilities_df.iloc[j][self.supply_col] / total_demand
            else:
                R_j[j] = 0
                
        facilities_df = facilities_df.copy()
        facilities_df['_R_j'] = R_j
        print(f"  Step1 氓庐聦忙聢聬: R_j 猫聦聝氓聸麓 [{R_j.min():.4f}, {R_j.max():.4f}]")
        return facilities_df, R_j
        
    def step2_accessibility(self, communities_df, facilities_df, od_matrix):
        """
        莽卢卢盲潞聦忙颅楼茂录職猫庐隆莽庐聴忙炉聫盲赂陋氓掳聫氓聦潞莽職聞氓聫炉猫戮戮忙聙搂 A_i
        
        A_i = sum(R_j) for all j where d_ij <= d_0
        
        氓聧鲁茂录職猫聬陆氓聹篓氓掳聫氓聦潞 i 忙颅楼猫隆聦猫聦聝氓聸麓氓聠聟忙聣聙忙聹聣猫庐戮忙聳陆莽職聞 R_j 盲鹿聥氓聮聦
        """
        R_j = facilities_df['_R_j'].values
        n_comm = len(communities_df)
        A_i = np.zeros(n_comm)
        
        for i in range(n_comm):
            reachable_mask = (od_matrix[i, :] <= self.search_radius) & np.isfinite(od_matrix[i, :])
            A_i[i] = R_j[reachable_mask].sum()
            
        communities_df = communities_df.copy()
        communities_df['A_i_2sfca'] = A_i
        
        # 氓聫炉猫戮戮忙聙搂忙聽聡氓聡聠氓聦聳茂录聢忙聳鹿盲戮驴猫路篓忙聦聡忙聽聡忙炉聰猫戮聝茂录聣
        A_max = A_i.max() if A_i.max() > 0 else 1
        communities_df['A_i_2sfca_norm'] = A_i / A_max
        
        print(f"  Step2 氓庐聦忙聢聬: A_i 猫聦聝氓聸麓 [{A_i.min():.4f}, {A_i.max():.4f}]")
        print(f"  忙聽聡氓聡聠氓聦聳 A_i 猫聦聝氓聸麓 [{communities_df['A_i_2sfca_norm'].min():.4f}, {communities_df['A_i_2sfca_norm'].max():.4f}]")
        return communities_df
        
    def fit_transform(self, communities_df, facilities_df, od_matrix):
        """氓庐聦忙聲麓莽職聞盲赂陇忙颅楼猫庐隆莽庐聴忙碌聛莽篓聥"""
        facilities_df, R_j = self.step1_supply_ratio(facilities_df, od_matrix)
        communities_df = self.step2_accessibility(communities_df, facilities_df, od_matrix)
        return communities_df, facilities_df
        

# 盲赂潞猫庐戮忙聳陆氓聢聠茅聟聧盲戮聸莽禄聶忙聺聝茅聡聧茂录聢忙篓隆忙聥聼氓陇搂盲录聴莽聜鹿猫炉聞猫炉聞氓聢聠莽職聞氓陆聮盲赂聙氓聦聳氓聙录茂录聣
if 'rating' not in poi_df.columns:
    poi_df['supply'] = np.random.uniform(0.5, 1.0, size=len(poi_df))  # 忙篓隆忙聥聼猫炉聞氓聢聠氓陆聮盲赂聙氓聦聳
else:
    poi_df['supply'] = poi_df['rating'].fillna(0.5) / 5.0

if 'population' not in communities_gdf.columns:
    communities_gdf['population'] = np.random.randint(500, 5000, size=len(communities_gdf))

print(f"猫庐戮忙聳陆忙聲掳忙聧庐: {len(poi_df)} 盲赂陋")
print(f"氓掳聫氓聦潞忙聲掳忙聧庐: {len(communities_gdf)} 盲赂陋")
```

```python
# ============================================================================
# 忙聻聞氓禄潞 OD 猫路聺莽娄禄莽聼漏茅聵碌茂录聢莽陇戮氓聦潞 芒聠聮 猫庐戮忙聳陆茂录聣
# ============================================================================

# 盲赂潞氓聤聽氓驴芦茅聙聼氓潞娄茂录聦氓聟聢盲陆驴莽聰篓氓掳聫猫搂聞忙篓隆忙碌聥猫炉聲
print("忙颅拢氓聹篓忙聻聞氓禄潞莽陇戮氓聦潞-猫庐戮忙聳陆 OD 猫路聺莽娄禄莽聼漏茅聵碌茂录聢猫路炉莽陆聭 Dijkstra茂录聣...")

START_TIME = time.time()
od_matrix = dist_calc.build_od_matrix_vectorized(
    communities_gdf[['center_lng', 'center_lat']].rename(
        columns={'center_lng': 'lng', 'center_lat': 'lat'}),
    poi_df[['lng', 'lat']],
    lng_col='lng', lat_col='lat'
)
ELAPSED = time.time() - START_TIME
print(f"OD 莽聼漏茅聵碌忙聻聞氓禄潞猫聙聴忙聴露: {ELAPSED:.1f}s")
print(f"莽聼漏茅聵碌氓陆垄莽聤露: {od_matrix.shape}")
print(f"忙聹聣忙聲聢猫路聺莽娄禄氓炉鹿: {np.sum(np.isfinite(od_matrix)):,} / {od_matrix.size:,} ({100*np.sum(np.isfinite(od_matrix))/od_matrix.size:.1f}%)")
```

```python
# ============================================================================
# 忙聣搂猫隆聦 2SFCA 猫庐隆莽庐聴茂录聢氓聢聠猫庐戮忙聳陆莽卤禄氓聻聥茂录聣
# ============================================================================

SEARCH_RADIUS_M = 1250  # 15 氓聢聠茅聮聼忙颅楼猫隆聦 = 1250m @ 5km/h

def run_2sfca_per_type(communities_gdf, poi_df, od_matrix, 
                       facility_type_col='facility_type',
                       search_radius=SEARCH_RADIUS_M):
    """氓炉鹿忙炉聫莽搂聧猫庐戮忙聳陆莽卤禄氓聻聥氓聢聠氓聢芦猫驴聬猫隆聦 2SFCA"""
    
    results = communities_gdf.copy()
    all_R_j = {}
    
    for ftype, group in poi_df.groupby(facility_type_col):
        if len(group) < 2:
            results[f'A_i_2sfca_{ftype}'] = 0
            results[f'A_i_2sfca_norm_{ftype}'] = 0
            continue
            
        # 猫聨路氓聫聳猫炉楼莽卤禄氓聻聥猫庐戮忙聳陆氓聹篓氓聨聼氓搂聥 poi_df 盲赂颅莽職聞氓聢聴莽麓垄氓录聲
        type_indices = poi_df[poi_df[facility_type_col] == ftype].index.tolist()
        type_od = od_matrix[:, type_indices]
        
        fac_group = group.copy().reset_index(drop=True)
        model = TwoStepFloatingCatchmentArea(
            search_radius_m=search_radius,
            supply_col='supply',
            demand_col='population'
        )
        
        comm_copy = results[['community_id', 'population']].copy()
        try:
            comm_result, fac_result = model.fit_transform(comm_copy, fac_group, type_od)
            # 氓掳聠莽禄聯忙聻聹氓炉鹿茅陆聬氓聬聢氓鹿露
            results = results.merge(
                comm_result[['community_id', 'A_i_2sfca', 'A_i_2sfca_norm']],
                on='community_id', how='left',
                suffixes=('', f'_{ftype}')
            )
            col = 'A_i_2sfca' if f'A_i_2sfca_{ftype}' not in results.columns else f'A_i_2sfca_{ftype}'
            if 'A_i_2sfca' in results.columns:
                results = results.rename(columns={
                    'A_i_2sfca': f'A_i_2sfca_{ftype}',
                    'A_i_2sfca_norm': f'A_i_2sfca_norm_{ftype}'
                })
            all_R_j[ftype] = fac_result['_R_j'].values
        except Exception as e:
            results[f'A_i_2sfca_{ftype}'] = 0
            results[f'A_i_2sfca_norm_{ftype}'] = 0
            print(f"  ! {ftype} 2SFCA 氓陇卤猫麓楼: {e}")
    
    # 莽禄录氓聬聢氓聫炉猫戮戮忙聙搂忙聦聡忙聲掳茂录聢忙聣聙忙聹聣猫庐戮忙聳陆莽卤禄氓聻聥氓聤聽忙聺聝氓鹿鲁氓聺聡茂录聣
    acc_cols = [c for c in results.columns if c.startswith('A_i_2sfca_norm_')]
    if acc_cols:
        results['A_i_2sfca_composite'] = results[acc_cols].mean(axis=1)
        results['A_i_2sfca_composite_raw'] = results[[c.replace('_norm', '') for c in acc_cols]].mean(axis=1)
    
    return results, all_R_j

print("忙聣搂猫隆聦 2SFCA茂录聢氓聢聠猫庐戮忙聳陆莽卤禄氓聻聥茂录聣...")
acc_results, R_j_dict = run_2sfca_per_type(
    communities_gdf, poi_df, od_matrix
)

# 莽禄聼猫庐隆忙聭聵猫娄聛
print("\n" + "=" * 60)
print("2SFCA 氓聫炉猫戮戮忙聙搂莽禄聯忙聻聹忙聭聵猫娄聛")
print("=" * 60)
acc_cols = [c for c in acc_results.columns if c.startswith('A_i_2sfca_norm_')]
for col in acc_cols:
    ftype = col.replace('A_i_2sfca_norm_', '')
    vals = acc_results[col].dropna()
    if len(vals) > 0:
        print(f"  {ftype:20s}: mean={vals.mean():.4f}, median={vals.median():.4f}, max={vals.max():.4f}")

print(f"\n莽禄录氓聬聢氓聫炉猫戮戮忙聙搂 (composite):")
comp = acc_results['A_i_2sfca_composite']
print(f"  mean={comp.mean():.4f}, median={comp.median():.4f}, std={comp.std():.4f}")
print(f"  盲陆聨氓聫炉猫戮戮忙聙搂(<0.2)氓掳聫氓聦潞: {(comp < 0.2).sum()} 盲赂陋")
print(f"  茅芦聵氓聫炉猫戮戮忙聙搂(>0.8)氓掳聫氓聦潞: {(comp > 0.8).sum()} 盲赂陋")
```

## 5. 茅芦聵忙聳炉猫隆掳氓聡聫忙聰鹿猫驴聸忙篓隆氓聻聥 (Gaussian 2SFCA)

### 5.1 忙篓隆氓聻聥氓聤篓忙聹潞

盲录聽莽禄聼 2SFCA 盲陆驴莽聰篓盲潞聦氓聙录猫路聺莽娄禄猫隆掳氓聡聫茂录聢氓聧聤氓戮聞氓聠聟=1茂录聦氓聧聤氓戮聞氓陇聳=0茂录聣茂录聦氓驴陆莽聲楼盲潞聠茂录職
1. 猫庐戮忙聳陆猫露聤猫驴聭茂录聦氓聫炉猫戮戮忙聙搂猫露聤茅芦聵茂录聢猫路聺莽娄禄忙聭漏忙聯娄忙聲聢氓潞聰茂录聣
2. 猫庐戮忙聳陆猫戮鹿莽聲聦氓陇聞氓聫炉猫戮戮忙聙搂莽職聞莽陋聛氓聫聵盲赂聧猫驴聻莽禄颅

**Gaussian 2SFCA**茂录聢Dai, 2010; Tao et al., 2020茂录聣莽聰篓茅芦聵忙聳炉猫隆掳氓聡聫忙聸驴盲禄拢盲潞聦氓聙录猫隆掳氓聡聫茂录職

$$G(d) = e^{-d^2 / 2\sigma^2} - e^{-d_0^2 / 2\sigma^2} \over 1 - e^{-d_0^2 / 2\sigma^2}$$

氓聟露盲赂颅 $\sigma$ 忙聨搂氓聢露猫隆掳氓聡聫茅聙聼氓潞娄茂录聢忙聹卢莽聽聰莽漏露氓聫聳 $d_0/3$茂录聣茂录聦
$d_0$ 盲赂潞忙聹聙氓陇搂忙聬聹莽麓垄氓聧聤氓戮聞茫聙聜

忙聰鹿猫驴聸氓聬聨莽職聞盲赂陇忙颅楼猫庐隆莽庐聴茂录職

$$R_j^{G} = \frac{S_j}{\sum_k P_k \cdot G(d_{kj})} \quad \quad A_i^{G} = \sum_j R_j^{G} \cdot G(d_{ij})$$

```python
# ============================================================================
# Gaussian 2SFCA 氓庐聻莽聨掳
# ============================================================================

class Gaussian2SFCA:
    """
    Gaussian 2SFCA茂录聢茅芦聵忙聳炉猫隆掳氓聡聫盲赂陇忙颅楼莽搂禄氓聤篓忙聬聹莽麓垄忙鲁聲茂录聣
    
    氓聫聜猫聙聝忙聳聡莽聦庐:
    - Dai, D. (2010). Racial/ethnic and socioeconomic disparities 
      in urban and regional planner access. Urban Studies.
    - Tao, Z., et al. (2020). Urban facility accessibility 
      based on modified 2SFCA. Environment and Planning B.
    """
    
    def __init__(self, search_radius_m=1250, sigma_ratio=1/3):
        self.d0 = search_radius_m
        self.sigma = search_radius_m * sigma_ratio  # sigma = d0/3
        
    def gaussian_weight(self, distance_m):
        """
        茅芦聵忙聳炉猫路聺莽娄禄猫隆掳氓聡聫忙聺聝茅聡聧
        G(d) 芒聢聢 [0,1], d=0 忙聴露 G=1, d=d0 忙聴露 G=0
        """
        if np.isinf(distance_m) or np.isnan(distance_m):
            return 0.0
        d = distance_m
        d0, sigma = self.d0, self.sigma
        if d >= d0:
            return 0.0
        G_d = math.exp(-d**2 / (2 * sigma**2))
        G_d0 = math.exp(-d0**2 / (2 * sigma**2))
        return (G_d - G_d0) / (1 - G_d0 + 1e-10)
        
    def fit_transform(self, communities_df, facilities_df, od_matrix):
        """
        忙聣搂猫隆聦 Gaussian 2SFCA 盲赂陇忙颅楼猫庐隆莽庐聴
        """
        n_comm = len(communities_df)
        n_fac = len(facilities_df)
        
        supply = facilities_df['supply'].values
        demand = communities_df['population'].values
        
        # 茅垄聞猫庐隆莽庐聴茅芦聵忙聳炉忙聺聝茅聡聧莽聼漏茅聵碌
        print(f"  茅垄聞猫庐隆莽庐聴茅芦聵忙聳炉忙聺聝茅聡聧莽聼漏茅聵碌 ({n_comm}脙聴{n_fac})...")
        w_od = np.zeros_like(od_matrix)
        for i in range(n_comm):
            for j in range(n_fac):
                w_od[i, j] = self.gaussian_weight(od_matrix[i, j])
        
        # Step 1: 猫庐隆莽庐聴 R_j^G
        R_j_G = np.zeros(n_fac)
        for j in range(n_fac):
            total_weighted_demand = 0
            for i in range(n_comm):
                if w_od[i, j] > 0:
                    total_weighted_demand += demand[i] * w_od[i, j]
            if total_weighted_demand > 0:
                R_j_G[j] = supply[j] / total_weighted_demand
        
        # Step 2: 猫庐隆莽庐聴 A_i^G
        A_i_G = np.zeros(n_comm)
        for i in range(n_comm):
            for j in range(n_fac):
                if w_od[i, j] > 0:
                    A_i_G[i] += R_j_G[j] * w_od[i, j]
        
        communities_df = communities_df.copy()
        communities_df['A_i_gaussian'] = A_i_G
        
        A_max = A_i_G.max() if A_i_G.max() > 0 else 1
        communities_df['A_i_gaussian_norm'] = A_i_G / A_max
        
        facilities_df = facilities_df.copy()
        facilities_df['_R_j_gaussian'] = R_j_G
        
        print(f"  Gaussian 2SFCA 氓庐聦忙聢聬: A_i^G 芒聢聢 [{A_i_G.min():.4f}, {A_i_G.max():.4f}]")
        return communities_df, facilities_df


# 氓炉鹿莽禄录氓聬聢忙聲掳忙聧庐猫驴聬猫隆聦 Gaussian 2SFCA
print("忙聣搂猫隆聦 Gaussian 2SFCA茂录聢莽庐聙氓聦聳莽聣聢茂录職猫庐戮忙聳陆芒聠聮莽禄录氓聬聢猫庐戮忙聳陆忙卤聽茂录聣...")

# 氓聢聸氓禄潞莽禄录氓聬聢猫庐戮忙聳陆忙卤聽茂录聢忙聣聙忙聹聣猫庐戮忙聳陆氓聬聢氓鹿露茂录聦supply=1茂录聣
pool_fac = poi_df[['facility_type', 'lng', 'lat']].copy()
pool_fac['supply'] = 1.0

# 忙聻聞氓禄潞莽陇戮氓聦潞芒聠聮猫庐戮忙聳陆忙卤聽莽職聞 OD 莽聼漏茅聵碌
od_pool = dist_calc.build_od_matrix_vectorized(
    communities_gdf[['center_lng', 'center_lat']].rename(
        columns={'center_lng': 'lng', 'center_lat': 'lat'}),
    pool_fac[['lng', 'lat']],
    lng_col='lng', lat_col='lat'
)

gaussian_model = Gaussian2SFCA(search_radius_m=SEARCH_RADIUS_M, sigma_ratio=1/3)
acc_results, pool_fac_result = gaussian_model.fit_transform(
    communities_gdf[['community_id', 'population']].copy(),
    pool_fac, od_pool
)

# 氓聬聢氓鹿露莽禄聯忙聻聹
acc_results = acc_results.merge(
    communities_gdf[['community_id', 'name', 'community_type', 
                      'center_lng', 'center_lat', 'population', 'geometry']],
    on='community_id', how='left'
)

print("\n" + "=" * 60)
print("Gaussian 2SFCA 莽禄聯忙聻聹忙聭聵猫娄聛")
print("=" * 60)
g_vals = acc_results['A_i_gaussian_norm']
print(f"忙聽聡氓聡聠氓聦聳氓聫炉猫戮戮忙聙搂: mean={g_vals.mean():.4f}, median={g_vals.median():.4f}, std={g_vals.std():.4f}")
print(f"盲陆聨氓聫炉猫戮戮忙聙搂(<0.2)氓掳聫氓聦潞: {(g_vals < 0.2).sum()} 盲赂陋 ({(g_vals < 0.2).mean()*100:.1f}%)")
print(f"茅芦聵氓聫炉猫戮戮忙聙搂(>0.8)氓掳聫氓聦潞: {(g_vals > 0.8).sum()} 盲赂陋 ({(g_vals > 0.8).mean()*100:.1f}%)")
```

## 6. 氓陇職忙聴露忙庐碌氓聫炉猫戮戮忙聙搂氓炉鹿忙炉聰氓聢聠忙聻聬茂录聢莽聶陆氓陇漏 / 氓陇聹茅聴麓茂录聣

### 6.1 忙聴露忙庐碌氓聢聮氓聢聠盲赂聨猫聬楼盲赂職莽潞娄忙聺聼

盲赂颅氓聸陆氓聼聨氓赂聜猫庐戮忙聳陆莽職聞猫聬楼盲赂職忙聴露茅聴麓氓路庐氓录聜忙聵戮猫聭聴氓陆卤氓聯聧氓聫炉猫戮戮忙聙搂茂录職
- **莽聶陆氓陇漏忙聴露忙庐碌** (08:00芒聙聯22:00): 氓陇搂茅聝篓氓聢聠猫庐戮忙聳陆忙颅拢氓赂赂猫聬楼盲赂職
- **氓陇聹茅聴麓忙聴露忙庐碌** (22:00芒聙聯08:00): 盲禄聟 24 氓掳聫忙聴露猫庐戮忙聳陆茂录聢盲戮驴氓聢漏氓潞聴茫聙聛茅聝篓氓聢聠猫聧炉氓潞聴茂录聣氓聫炉氓聫聤

忙聹卢莽聽聰莽漏露氓掳聠莽聶陆氓陇漏/氓陇聹茅聴麓氓聫炉猫戮戮忙聙搂氓路庐氓录聜氓庐職盲鹿聣盲赂潞 **忙聴露茅聴麓猫麓芦氓聸掳忙聦聡忙聲掳 (Temporal Poverty Index, TPI)**茂录職

$$TPI_i = \frac{A_i^{night} - A_i^{day}}{A_i^{day}} \times 100\%$$

```python
# ============================================================================
# 忙聴露忙庐碌猫聬楼盲赂職莽潞娄忙聺聼盲赂聨氓陇職忙聴露忙庐碌 2SFCA
# ============================================================================

FACILITY_NIGHT_SERVICE = {
    'convenience': 1.0,    # 盲戮驴氓聢漏氓潞聴 24h
    'pharmacy': 0.3,       # 茅聝篓氓聢聠猫聧炉氓潞聴 24h
    'hospital': 0.2,       # 氓聦禄茅聶垄忙聙楼猫炉聤
    'clinic': 0.05,       # 猫炉聤忙聣聙氓陇聹茅聴麓氓掳聭
    'supermarket': 0.1,    # 茅聝篓氓聢聠猫露聟氓赂聜 24h
    'bank': 0.0,          # 茅聯露猫隆聦氓陇聹茅聴麓氓聟鲁茅聴颅
    'atm': 1.0,          # ATM 24h
    'school': 0.0,       # 氓颅娄忙聽隆氓陇聹茅聴麓氓聟鲁茅聴颅
    'kindergarten': 0.0,   # 氓鹿录氓聞驴氓聸颅氓陇聹茅聴麓氓聟鲁茅聴颅
    'bus_stop': 1.0,     # 氓聟卢盲潞陇莽芦聶
}

def run_multi_period_2sfca(communities_df, poi_df, od_matrix,
                            search_radius=SEARCH_RADIUS_M):
    """
    氓聢聠氓聢芦猫庐隆莽庐聴莽聶陆氓陇漏氓聮聦氓陇聹茅聴麓莽職聞 2SFCA 氓聫炉猫戮戮忙聙搂
    """
    results = communities_df.copy()
    
    for period, night_multiplier in [('day', 1.0), ('night', None)]:
        print(f"\n氓陇聞莽聬聠忙聴露忙庐碌: {period}")
        
        # 忙聽鹿忙聧庐忙聴露忙庐碌莽颅聸茅聙聣猫庐戮忙聳陆
        if period == 'night':
            # 氓陇聹茅聴麓: 盲禄聟盲驴聺莽聲聶氓陇聹茅聴麓忙聹聧氓聤隆猫庐戮忙聳陆
            mask = poi_df['facility_type'].map(
                lambda t: FACILITY_NIGHT_SERVICE.get(t, 0) > 0
            )
            period_poi = poi_df[mask].copy()
            if len(period_poi) == 0:
                results[f'A_i_2sfca_{period}'] = 0.0
                results[f'A_i_2sfca_norm_{period}'] = 0.0
                continue
            # 氓潞聰莽聰篓氓陇聹茅聴麓忙聹聧氓聤隆氓聫炉莽聰篓忙聙搂莽鲁禄忙聲掳
            period_poi['effective_supply'] = period_poi['supply'] * period_poi['facility_type'].map(
                lambda t: FACILITY_NIGHT_SERVICE.get(t, 0)
            )
            period_poi_index = period_poi.index.tolist()
            period_od = od_matrix[:, period_poi_index]
        else:
            period_poi = poi_df.copy()
            period_poi['effective_supply'] = period_poi['supply']
            period_od = od_matrix
        
        # 2SFCA
        model = TwoStepFloatingCatchmentArea(
            search_radius_m=search_radius,
            supply_col='effective_supply',
            demand_col='population'
        )
        
        comm_tmp = results[['community_id', 'population']].copy()
        fac_tmp = period_poi[['supply', 'effective_supply']].reset_index(drop=True)
        fac_tmp.index = period_poi.index
        fac_tmp = period_poi[['effective_supply']].copy()
        fac_tmp['supply'] = period_poi['effective_supply']
        
        comm_result, fac_result = model.fit_transform(comm_tmp, fac_tmp, period_od)
        
        results = results.merge(
            comm_result[['community_id', 'A_i_2sfca', 'A_i_2sfca_norm']],
            on='community_id', how='left',
            suffixes=('', f'_{period}')
        )
        
        col_name = 'A_i_2sfca'
        norm_col = 'A_i_2sfca_norm'
        results = results.rename(columns={
            col_name: f'A_i_2sfca_{period}',
            norm_col: f'A_i_2sfca_norm_{period}'
        })
    
    # 忙聴露茅聴麓猫麓芦氓聸掳忙聦聡忙聲掳 TPI
    day_vals = results['A_i_2sfca_norm_day'].fillna(0)
    night_vals = results['A_i_2sfca_norm_night'].fillna(0)
    
    results['TPI'] = np.where(
        day_vals > 0,
        (night_vals - day_vals) / day_vals * 100,
        0
    )
    
    # 氓聫炉猫戮戮忙聙搂氓聫聵氓聦聳茂录聢莽禄聺氓炉鹿氓聙录茂录聣
    results['accessibility_gap'] = day_vals - night_vals
    
    print(f"\n莽聶陆氓陇漏氓聫炉猫戮戮忙聙搂: mean={day_vals.mean():.4f}, median={day_vals.median():.4f}")
    print(f"氓陇聹茅聴麓氓聫炉猫戮戮忙聙搂: mean={night_vals.mean():.4f}, median={night_vals.median():.4f}")
    print(f"TPI 忙聴露茅聴麓猫麓芦氓聸掳忙聦聡忙聲掳: mean={results['TPI'].mean():.1f}%, 忙聹聙氓陇搂={results['TPI'].max():.1f}%")
    
    return results

acc_results = run_multi_period_2sfca(
    acc_results, poi_df, od_matrix, search_radius=SEARCH_RADIUS_M
)
```

**芒聭隆 氓聫炉猫戮戮忙聙搂氓聣楼氓陇潞忙聦聡忙聲掳 (Accessibility Deprivation Index, ADI)**茂录職氓聙聼茅聣麓猫聥卤氓聸陆 Indices of Multiple Deprivation茂录聦氓掳聠氓聫炉猫戮戮忙聙搂茅聡聫氓聦聳猫陆卢忙聧垄盲赂潞"猫垄芦氓聣楼氓陇潞莽篓聥氓潞娄"

**芒聭垄 氓聢聠盲陆聧忙聲掳氓炉鹿忙炉聰氓聢聠忙聻聬**茂录職忙聹聙茅芦聵/忙聹聙盲陆聨20%氓掳聫氓聦潞莽職聞氓聫炉猫戮戮忙聙搂忙炉聰氓聙录茂录聦忙聫颅莽陇潞氓路庐猫路聺莽職聞莽聹聼氓庐聻猫搂聞忙篓隆

```python
# ============================================================================
# 氓聟卢氓鹿鲁忙聙搂氓聢聠忙聻聬茂录職Gini莽鲁禄忙聲掳茫聙聛忙麓聸盲录娄氓聟鹿忙聸虏莽潞驴盲赂聨氓聫炉猫戮戮忙聙搂氓聣楼氓陇潞忙聦聡忙聲掳
# ============================================================================

print("=" * 70)
print("氓聟卢氓鹿鲁忙聙搂氓聢聠忙聻聬 芒聙聰 氓聫炉猫戮戮忙聙搂氓聢聠茅聟聧莽職聞氓聟卢忙颅拢忙聙搂忙拢聙茅陋聦")
print("=" * 70)

def compute_gini(values):
    """猫庐隆莽庐聴氓聼潞氓掳录莽鲁禄忙聲掳茂录聢Gini coefficient茂录聣"""
    values = np.array(values).flatten()
    values = values[~np.isnan(values)]
    if len(values) == 0:
        return np.nan
    values = np.sort(values)
    n = len(values)
    mean_val = np.mean(values)
    if mean_val == 0:
        return np.nan
    index = np.arange(1, n + 1)
    gini = (2 * np.sum(index * values) - (n + 1) * np.sum(values)) / (n * np.sum(values))
    return gini

def lorenz_curve(values):
    """猫庐隆莽庐聴忙麓聸盲录娄氓聟鹿忙聸虏莽潞驴忙聲掳忙聧庐"""
    values = np.sort(values.flatten())
    values = values[~np.isnan(values)]
    cum_share = np.cumsum(values) / np.sum(values)
    pop_share = np.arange(1, len(values) + 1) / len(values)
    return pop_share, cum_share

def compute_deprivation_index(accessibility_values):
    """氓聫炉猫戮戮忙聙搂氓聣楼氓陇潞忙聦聡忙聲掳 ADI = 1 - A_i / A_max"""
    A_max = np.nanmax(accessibility_values)
    if A_max == 0:
        return np.full_like(accessibility_values, np.nan)
    return 1 - accessibility_values / A_max

# 芒聙聰芒聙聰芒聙聰芒聙聰芒聙聰芒聙聰芒聙聰芒聙聰芒聙聰芒聙聰芒聙聰芒聙聰芒聙聰芒聙聰芒聙聰芒聙聰芒聙聰芒聙聰芒聙聰芒聙聰芒聙聰芒聙聰芒聙聰芒聙聰芒聙聰芒聙聰
# 1. 氓聬聢氓鹿露猫聞聠氓录卤忙聙搂盲赂聨氓聫炉猫戮戮忙聙搂忙聲掳忙聧庐
# 芒聙聰芒聙聰芒聙聰芒聙聰芒聙聰芒聙聰芒聙聰芒聙聰芒聙聰芒聙聰芒聙聰芒聙聰芒聙聰芒聙聰芒聙聰芒聙聰芒聙聰芒聙聰芒聙聰芒聙聰芒聙聰芒聙聰芒聙聰芒聙聰芒聙聰芒聙聰
if "MVI" not in acc_results.columns:
    cols_to_merge = ["community_id", "HV", "SEV", "SAV", "PV", "MVI",
                      "is_vulnerability_stacked", "community_type"]
    acc_results = acc_results.merge(
        communities_gdf[cols_to_merge], on="community_id", how="left", suffixes=("", "_c")
    )

# 猫庐隆莽庐聴氓聣楼氓陇潞忙聦聡忙聲掳
day_vals = acc_results.get("A_i_2sfca_norm_day", pd.Series([np.nan]*len(acc_results))).fillna(0).values
night_vals = acc_results.get("A_i_2sfca_norm_night", pd.Series([np.nan]*len(acc_results))).fillna(0).values
gauss_vals = acc_results.get("A_i_gaussian_norm", pd.Series([np.nan]*len(acc_results))).fillna(0).values

acc_results["ADI_day"] = compute_deprivation_index(day_vals)
acc_results["ADI_night"] = compute_deprivation_index(night_vals)
acc_results["ADI_gaussian"] = compute_deprivation_index(gauss_vals)

# 芒聙聰芒聙聰芒聙聰芒聙聰芒聙聰芒聙聰芒聙聰芒聙聰芒聙聰芒聙聰芒聙聰芒聙聰芒聙聰芒聙聰芒聙聰芒聙聰芒聙聰芒聙聰芒聙聰芒聙聰芒聙聰芒聙聰芒聙聰芒聙聰芒聙聰芒聙聰
# 2. Gini 莽鲁禄忙聲掳猫庐隆莽庐聴
# 芒聙聰芒聙聰芒聙聰芒聙聰芒聙聰芒聙聰芒聙聰芒聙聰芒聙聰芒聙聰芒聙聰芒聙聰芒聙聰芒聙聰芒聙聰芒聙聰芒聙聰芒聙聰芒聙聰芒聙聰芒聙聰芒聙聰芒聙聰芒聙聰芒聙聰芒聙聰
print("\n茫聙聬Gini 莽鲁禄忙聲掳 芒聙聰 氓聫炉猫戮戮忙聙搂氓聢聠茅聟聧氓聟卢氓鹿鲁忙聙搂茫聙聭")
print("-" * 70)

gini_results = {}
for col, label in [("A_i_2sfca_norm_day", "莽聶陆氓陇漏2SFCA"),
                    ("A_i_2sfca_norm_night", "氓陇聹茅聴麓2SFCA"),
                    ("A_i_gaussian_norm", "Gaussian 2SFCA")]:
    if col in acc_results.columns:
        vals = acc_results[col].fillna(0).values
        gini = compute_gini(vals)
        gini_results[label] = gini
        interp = "茅芦聵氓潞娄氓聟卢氓鹿鲁" if gini < 0.2 else "莽聸赂氓炉鹿氓聟卢氓鹿鲁" if gini < 0.35 else "盲赂聧氓鹿鲁莽颅聣" if gini < 0.5 else "忙聻聛氓潞娄盲赂聧氓鹿鲁莽颅聣"
        print(f"  {label:15s} Gini = {gini:.4f}  [{interp}]")

print("\n猫搂拢猫炉禄茂录職")
print(f"  脗路 莽聶陆氓陇漏氓聫炉猫戮戮忙聙搂 Gini = {gini_results.get("莽聶陆氓陇漏2SFCA", 0):.4f}")
print("  脗路 猫聥楼 Gini > 0.4茂录聦猫炉麓忙聵聨15氓聢聠茅聮聼莽聰聼忙麓禄氓聹聢莽職聞猫碌聞忙潞聬氓聢聠茅聟聧氓颅聵氓聹篓忙聵戮猫聭聴莽漏潞茅聴麓盲赂聧氓聟卢氓鹿鲁")
print(f"  脗路 氓陇聹茅聴麓氓聫炉猫戮戮忙聙搂 Gini = {gini_results.get("氓陇聹茅聴麓2SFCA", 0):.4f}")
print("  脗路 氓陇聹茅聴麓盲赂聧氓鹿鲁莽颅聣莽篓聥氓潞娄茅聙職氓赂赂茅芦聵盲潞聨莽聶陆氓陇漏茂录聦氓聫聧忙聵聽24h猫庐戮忙聳陆莽職聞莽篓聙莽录潞忙聙搂")

# 芒聙聰芒聙聰芒聙聰芒聙聰芒聙聰芒聙聰芒聙聰芒聙聰芒聙聰芒聙聰芒聙聰芒聙聰芒聙聰芒聙聰芒聙聰芒聙聰芒聙聰芒聙聰芒聙聰芒聙聰芒聙聰芒聙聰芒聙聰芒聙聰芒聙聰芒聙聰
# 3. 忙麓聸盲录娄氓聟鹿忙聸虏莽潞驴氓聫炉猫搂聠氓聦聳
# 芒聙聰芒聙聰芒聙聰芒聙聰芒聙聰芒聙聰芒聙聰芒聙聰芒聙聰芒聙聰芒聙聰芒聙聰芒聙聰芒聙聰芒聙聰芒聙聰芒聙聰芒聙聰芒聙聰芒聙聰芒聙聰芒聙聰芒聙聰芒聙聰芒聙聰芒聙聰
fig, axes = plt.subplots(1, 2, figsize=(16, 6))

ax1 = axes[0]
ax1.set_title("忙麓聸盲录娄氓聟鹿忙聸虏莽潞驴 芒聙聰 氓聫炉猫戮戮忙聙搂氓聢聠茅聟聧氓聟卢氓鹿鲁忙聙搂", fontsize=13, fontweight="bold")
ax1.plot([0, 1], [0, 1], "k--", linewidth=2, label="莽禄聺氓炉鹿氓鹿鲁莽颅聣莽潞驴 (G=0)")

colors = {"莽聶陆氓陇漏2SFCA": "#3498db", "氓陇聹茅聴麓2SFCA": "#e74c3c", "Gaussian 2SFCA": "#27ae60"}
for label, col_name in [("莽聶陆氓陇漏2SFCA", "A_i_2sfca_norm_day"),
                          ("氓陇聹茅聴麓2SFCA", "A_i_2sfca_norm_night"),
                          ("Gaussian 2SFCA", "A_i_gaussian_norm")]:
    if col_name in acc_results.columns:
        vals = acc_results[col_name].fillna(0).values
        x, y = lorenz_curve(vals)
        g = compute_gini(vals)
        ax1.plot(x, y, linewidth=2.5, label=f"{label} (G={g:.3f})", color=colors.get(label, "gray"))
        ax1.fill_between(x, y, x, alpha=0.1, color=colors.get(label, "gray"))

ax1.set_xlabel("盲潞潞氓聫拢莽麓炉莽搂炉忙炉聰盲戮聥", fontsize=11)
ax1.set_ylabel("氓聫炉猫戮戮忙聙搂莽麓炉莽搂炉忙炉聰盲戮聥", fontsize=11)
ax1.legend(fontsize=10, loc="upper left")
ax1.set_xlim(0, 1)
ax1.set_ylim(0, 1)
ax1.grid(True, alpha=0.3)

# 芒聙聰芒聙聰芒聙聰芒聙聰芒聙聰芒聙聰芒聙聰芒聙聰芒聙聰芒聙聰芒聙聰芒聙聰芒聙聰芒聙聰芒聙聰芒聙聰芒聙聰芒聙聰芒聙聰芒聙聰芒聙聰芒聙聰芒聙聰芒聙聰芒聙聰芒聙聰
# 4. 氓聫炉猫戮戮忙聙搂氓聣楼氓陇潞忙聦聡忙聲掳氓聢聠忙聻聬
# 芒聙聰芒聙聰芒聙聰芒聙聰芒聙聰芒聙聰芒聙聰芒聙聰芒聙聰芒聙聰芒聙聰芒聙聰芒聙聰芒聙聰芒聙聰芒聙聰芒聙聰芒聙聰芒聙聰芒聙聰芒聙聰芒聙聰芒聙聰芒聙聰芒聙聰芒聙聰
ax2 = axes[1]
adi_col = "ADI_gaussian"
type_cn = {"urban_village": "氓聼聨盲赂颅忙聺聭", "affordable_housing": "盲驴聺茅職聹忙聢驴",
            "commodity_housing": "氓聲聠氓聯聛忙聢驴", "high_end": "茅芦聵莽芦炉莽陇戮氓聦潞"}
type_colors2 = {"urban_village": "#c0392b", "affordable_housing": "#e67e22",
                 "commodity_housing": "#27ae60", "high_end": "#2980b9"}

for ctype in type_colors2:
    subset = acc_results[acc_results["community_type"] == ctype]
    if len(subset) == 0:
        continue
    vals = subset[adi_col].dropna().sort_values()
    if len(vals) == 0:
        continue
    x_vals = np.linspace(0, 1, len(vals))
    label = type_cn.get(ctype, ctype)
    color = type_colors2[ctype]
    ax2.plot(x_vals, vals.values, linewidth=2.5, label=f"{label} (n={len(vals)})", color=color)
    ax2.fill_between(x_vals, vals.values, alpha=0.05, color=color)

ax2.set_xlabel("氓掳聫氓聦潞莽麓炉莽搂炉忙炉聰盲戮聥茂录聢忙聦聣氓聣楼氓陇潞莽篓聥氓潞娄忙聨聮氓潞聫茂录聣", fontsize=11)
ax2.set_ylabel("氓聫炉猫戮戮忙聙搂氓聣楼氓陇潞忙聦聡忙聲掳 (ADI)", fontsize=11)
ax2.set_title("氓聫炉猫戮戮忙聙搂氓聣楼氓陇潞忙聸虏莽潞驴 芒聙聰 猫掳聛猫垄芦氓聣楼氓陇潞氓戮聴忙聹聙盲赂楼茅聡聧茂录聼", fontsize=13, fontweight="bold")
ax2.legend(fontsize=10)
ax2.grid(True, alpha=0.3)

plt.tight_layout()
plt.savefig(os.path.join(BASE_DIR, "06_equity_analysis.png"), dpi=150, bbox_inches="tight", facecolor="white")
plt.show()
print("\n氓聸戮猫隆篓氓路虏盲驴聺氓颅聵: 06_equity_analysis.png")

# 芒聙聰芒聙聰芒聙聰芒聙聰芒聙聰芒聙聰芒聙聰芒聙聰芒聙聰芒聙聰芒聙聰芒聙聰芒聙聰芒聙聰芒聙聰芒聙聰芒聙聰芒聙聰芒聙聰芒聙聰芒聙聰芒聙聰芒聙聰芒聙聰芒聙聰芒聙聰
# 5. 氓聢聠莽戮陇盲陆聯氓聣楼氓陇潞氓炉鹿忙炉聰莽禄聼猫庐隆
# 芒聙聰芒聙聰芒聙聰芒聙聰芒聙聰芒聙聰芒聙聰芒聙聰芒聙聰芒聙聰芒聙聰芒聙聰芒聙聰芒聙聰芒聙聰芒聙聰芒聙聰芒聙聰芒聙聰芒聙聰芒聙聰芒聙聰芒聙聰芒聙聰芒聙聰芒聙聰
print("\n" + "=" * 70)
print("茫聙聬氓聟鲁茅聰庐氓聫聭莽聨掳茫聙聭盲赂聧氓聬聦莽戮陇盲陆聯氓聫炉猫戮戮忙聙搂氓聣楼氓陇潞氓炉鹿忙炉聰")
print("=" * 70)

equity_summary = []
for ctype, cname in [("urban_village", "氓聼聨盲赂颅忙聺聭"), ("affordable_housing", "盲驴聺茅職聹忙聢驴"),
                       ("commodity_housing", "氓聲聠氓聯聛忙聢驴"), ("high_end", "茅芦聵莽芦炉莽陇戮氓聦潞")]:
    subset = acc_results[acc_results["community_type"] == ctype]
    if len(subset) == 0:
        continue
    acc_day = subset.get("A_i_2sfca_norm_day", pd.Series([np.nan]*len(subset))).dropna()
    acc_night = subset.get("A_i_2sfca_norm_night", pd.Series([np.nan]*len(subset))).dropna()
    acc_g = subset.get("A_i_gaussian_norm", pd.Series([np.nan]*len(subset))).dropna()
    equity_summary.append({
        "莽戮陇盲陆聯": cname,
        "氓掳聫氓聦潞忙聲掳": len(subset),
        "莽聶陆氓陇漏氓聫炉猫戮戮忙聙搂氓聺聡氓聙录": acc_day.mean() if len(acc_day) > 0 else np.nan,
        "氓陇聹茅聴麓氓聫炉猫戮戮忙聙搂氓聺聡氓聙录": acc_night.mean() if len(acc_night) > 0 else np.nan,
        "莽禄录氓聬聢氓聫炉猫戮戮忙聙搂氓聺聡氓聙录": acc_g.mean() if len(acc_g) > 0 else np.nan,
        "ADI氓聺聡氓聙录": subset["ADI_day"].mean() if "ADI_day" in subset.columns else np.nan,
    })

equity_df = pd.DataFrame(equity_summary)
print(equity_df.to_string(index=False))

# 芒聙聰芒聙聰芒聙聰芒聙聰芒聙聰芒聙聰芒聙聰芒聙聰芒聙聰芒聙聰芒聙聰芒聙聰芒聙聰芒聙聰芒聙聰芒聙聰芒聙聰芒聙聰芒聙聰芒聙聰芒聙聰芒聙聰芒聙聰芒聙聰芒聙聰芒聙聰
# 6. 氓聫聦茅聡聧氓聣楼氓陇潞猫炉聠氓聢芦
# 芒聙聰芒聙聰芒聙聰芒聙聰芒聙聰芒聙聰芒聙聰芒聙聰芒聙聰芒聙聰芒聙聰芒聙聰芒聙聰芒聙聰芒聙聰芒聙聰芒聙聰芒聙聰芒聙聰芒聙聰芒聙聰芒聙聰芒聙聰芒聙聰芒聙聰芒聙聰
print("\n" + "-" * 70)
print("茫聙聬氓聫聦茅聡聧氓聣楼氓陇潞 (Double Deprivation) 猫炉聠氓聢芦茫聙聭")
print("-" * 70)

if "MVI" in acc_results.columns and "ADI_day" in acc_results.columns:
    acc_results["double_deprived"] = (
        (acc_results["MVI"] > 0.5) & (acc_results["ADI_day"] > 0.5)
    )
    dd_count = acc_results["double_deprived"].sum()
    dd_total = len(acc_results)
    print(f"  氓聫聦茅聡聧氓聣楼氓陇潞氓掳聫氓聦潞忙聲掳茅聡聫: {dd_count} / {dd_total} ({dd_count/dd_total*100:.1f}%)")
    dd_communities = acc_results[acc_results["double_deprived"]]
    if len(dd_communities) > 0:
        print("\n  氓聫聦茅聡聧氓聣楼氓陇潞氓掳聫氓聦潞猫炉娄忙聝聟茂录聢茅芦聵猫聞聠氓录卤忙聙搂 + 盲陆聨氓聫炉猫戮戮忙聙搂茂录聣:")
        cols_show = ["name", "community_type", "MVI", "ADI_day"]
        available = [c for c in cols_show if c in dd_communities.columns]
        display_df = dd_communities[available].copy()
        display_df["莽卤禄氓聻聥"] = display_df["community_type"].map(type_cn)
        print(display_df[["name", "莽卤禄氓聻聥", "MVI", "ADI_day"]].head(10).to_string(index=False))

    valid_mask = ~(acc_results["MVI"].isna() | acc_results["ADI_day"].isna())
    if valid_mask.sum() > 10:
        corr = acc_results.loc[valid_mask, "MVI"].corr(acc_results.loc[valid_mask, "ADI_day"])
        print(f"  MVI 盲赂聨 ADI 莽聸赂氓聟鲁莽鲁禄忙聲掳: r = {corr:.4f}")
        if corr > 0.3:
            print("  猫搂拢猫炉禄: 忙颅拢莽聸赂氓聟鲁忙聵戮猫聭聴 芒聠聮 猫聞聠氓录卤忙聙搂猫露聤茅芦聵莽職聞氓掳聫氓聦潞茂录聦猫垄芦氓聣楼氓陇潞莽篓聥氓潞娄猫露聤茅芦聵茂录聢莽漏潞茅聴麓盲赂聧氓聟卢氓鹿鲁茂录聣")
        else:
            print("  猫搂拢猫炉禄: 莽聸赂氓聟鲁忙聙搂猫戮聝氓录卤 芒聠聮 猫聞聠氓录卤忙聙搂盲赂聨氓聫炉猫戮戮忙聙搂氓聟鲁莽鲁禄猫戮聝盲赂潞氓陇聧忙聺聜")

print("\n" + "=" * 70)
print("茫聙聬盲潞潞忙聳聡氓聫聧忙聙聺茫聙聭忙聲掳氓颅聴猫聝聦氓聬聨莽職聞氓聟卢氓鹿鲁忙聙搂氓聧卤忙聹潞")
print("=" * 70)
print("""
氓陆聯忙聢聭盲禄卢猫庐隆莽庐聴 Gini 莽鲁禄忙聲掳忙聴露茂录聦忙聲掳氓颅聴猫聝聦氓聬聨忙聵炉莽聹聼氓庐聻莽職聞盲潞潞莽聰聼茂录職

  脗路 氓聼聨盲赂颅忙聺聭氓卤聟忙掳聭莽職聞氓鹿鲁氓聺聡氓聫炉猫戮戮忙聙搂茂录聦氓戮聙氓戮聙盲赂聧氓聢掳茅芦聵莽芦炉莽陇戮氓聦潞莽職聞1/3
  脗路 盲赂聙盲陆聧盲陆聫氓聹篓氓聼聨盲赂颅忙聺聭莽職聞猫聙聛盲潞潞茂录聦氓陇聹茅聴麓莽聰聼莽聴聟忙聴露忙聹聙猫驴聭莽職聞24h猫聧炉氓潞聴氓聫炉猫聝陆茅聹聙猫娄聛忙颅楼猫隆聦25氓聢聠茅聮聼
  脗路 猫驴聶盲赂聧忙聵炉"15氓聢聠茅聮聼氓聼聨氓赂聜"茂录聦猫驴聶忙聵炉"25氓聢聠茅聮聼氓聸掳氓卤聙"

  忙聰驴莽颅聳氓聬芦盲鹿聣茂录職
  1. 氓鹿鲁氓聺聡氓聫炉猫戮戮忙聙搂猫戮戮忙聽聡 芒聣聽 忙聣聙忙聹聣莽戮陇盲陆聯氓聫炉猫戮戮忙聙搂猫戮戮忙聽聡
  2. 茅聹聙猫娄聛"氓路庐氓录聜氓聦聳莽職聞"15氓聢聠茅聮聼莽聰聼忙麓禄氓聹聢猫搂聞氓聢聮芒聙聰芒聙聰氓炉鹿氓录卤氓聤驴莽陇戮氓聦潞忙聤聲氓聟楼忙聸麓氓陇職猫碌聞忙潞聬
  3. Gini莽鲁禄忙聲掳忙聵炉莽聸聭忙碌聥莽漏潞茅聴麓氓聟卢氓鹿鲁忙聙搂莽職聞氓聟鲁茅聰庐茅垄聞猫颅娄忙聦聡忙聽聡
""")
```

```python
# ============================================================================
# 莽漏潞茅聴麓猫聡陋莽聸赂氓聟鲁氓聢聠忙聻聬 (Moran's I & LISA)
# ============================================================================

from libpysal.weights import Queen, KNN, DistanceBand
from esda.moran import Moran, Moran_Local
import libpysal as lps

# 猫陆卢忙聧垄盲赂潞 GeoDataFrame茂录聢莽聰篓盲潞聨莽漏潞茅聴麓忙聺聝茅聡聧猫庐隆莽庐聴茂录聣
acc_gdf = gpd.GeoDataFrame(acc_results, geometry='geometry', crs='EPSG:4326')
acc_gdf = acc_gdf.dropna(subset=['center_lng', 'center_lat'])
acc_gdf = acc_gdf[acc_gdf.geometry.is_valid].copy()
acc_gdf = acc_gdf.reset_index(drop=True)

print(f"忙聹聣忙聲聢氓掳聫氓聦潞茂录聢莽聰篓盲潞聨莽漏潞茅聴麓氓聢聠忙聻聬茂录聣: {len(acc_gdf)} 盲赂陋")

# 忙聻聞氓禄潞莽漏潞茅聴麓忙聺聝茅聡聧莽聼漏茅聵碌茂录聢Queen 茅聜禄忙聨楼 + KNN 猫隆楼氓聟聟茂录聣
print("忙聻聞氓禄潞莽漏潞茅聴麓忙聺聝茅聡聧莽聼漏茅聵碌...")
try:
    w_queen = Queen.from_dataframe(acc_gdf, use_index=False)
    w_queen.transform = 'r'  # 猫隆聦忙聽聡氓聡聠氓聦聳
    print(f"  Queen 茅聜禄忙聨楼: {len(w_queen.neighbors)} 盲赂陋茅聜禄氓卤聟氓聟鲁莽鲁禄")
except Exception as e:
    print(f"  Queen 茅聜禄忙聨楼氓陇卤猫麓楼: {e}")
    w_queen = None

# 盲陆驴莽聰篓 KNN 忙聺聝茅聡聧莽聼漏茅聵碌茂录聢k=8茂录聦忙聹聙氓赂赂莽聰篓猫庐戮氓庐職茂录聣
coords = np.array(list(zip(acc_gdf.geometry.centroid.x, acc_gdf.geometry.centroid.y)))
w_knn = KNN.from_dataframe(acc_gdf, k=8)
w_knn.transform = 'r'
print(f"  KNN 忙聺聝茅聡聧 (k=8): {len(w_knn.neighbors)} 盲赂陋茅聜禄氓卤聟氓聟鲁莽鲁禄")

# 氓聬聢氓鹿露盲赂陇莽搂聧忙聺聝茅聡聧茂录聢氓聫聳氓鹿鲁氓聺聡茂录聣
# 忙聹卢莽聽聰莽漏露盲陆驴莽聰篓 KNN 忙聺聝茅聡聧茂录聢忙聸麓莽篓鲁氓庐職茂录聣
w_use = w_knn
```

```python
# ============================================================================
# 氓聟篓氓卤聙 Moran's I 忙拢聙茅陋聦
# ============================================================================

print("\n" + "=" * 60)
print("氓聟篓氓卤聙莽漏潞茅聴麓猫聡陋莽聸赂氓聟鲁忙拢聙茅陋聦 (Moran's I)")
print("=" * 60)

moran_results = {}
for period in ['day', 'night', 'gaussian_norm']:
    col = f'A_i_2sfca_norm_{period}' if period in ['day', 'night'] else f'A_i_{period}'
    
    if col not in acc_gdf.columns:
        continue
    
    y = acc_gdf[col].values
    y = np.nan_to_num(y, nan=np.nanmean(y))
    
    try:
        # 氓聟篓氓卤聙 Moran's I
        moran = Moran(y, w_use, permutations=999)
        moran_results[period] = {
            'I': moran.I,
            'E_I': moran.EI,
            'p_z': moran.p_z,
            'p_sim': moran.p_sim
        }
        sig = '***' if moran.p_z < 0.001 else '**' if moran.p_z < 0.01 else '*' if moran.p_z < 0.05 else 'ns'
        print(f"\n{period.upper()} 忙聴露忙庐碌氓聫炉猫戮戮忙聙搂:")
        print(f"  Moran's I    = {moran.I:.4f}")
        print(f"  E[I]         = {moran.EI:.4f}")
        print(f"  z-score      = {(moran.I - moran.EI)/np.sqrt(moran.VI_norm):.3f}")
        print(f"  p-value      = {moran.p_z:.4f} {sig}")
        print(f"  莽漏潞茅聴麓忙聽录氓卤聙     : {'忙聵戮猫聭聴猫聛職茅聸聠' if moran.p_z < 0.05 and moran.I > 0 else '忙聵戮猫聭聴氓聢聠忙聲拢' if moran.p_z < 0.05 and moran.I < 0 else '莽漏潞茅聴麓茅職聫忙聹潞'}")
    except Exception as e:
        print(f"  ! {period} Moran's I 氓陇卤猫麓楼: {e}")

# 氓聫炉猫戮戮忙聙搂忙聽录氓卤聙氓聫炉猫搂聠氓聦聳茂录聢Moran 忙聲拢莽聜鹿氓聸戮茂录聣
fig, axes = plt.subplots(1, 2, figsize=(14, 6))

for idx, period in enumerate(['day', 'night']):
    col = f'A_i_2sfca_norm_{period}'
    y = acc_gdf[col].fillna(acc_gdf[col].mean()).values
    
    lag = lps.weights.lag_spatial(w_use, y)
    ax = axes[idx]
    
    # 忙聽聡氓聡聠氓聦聳
    y_std = (y - y.mean()) / y.std()
    lag_std = (lag - lag.mean()) / lag.std()
    
    # 猫卤隆茅聶聬茅垄聹猫聣虏
    colors = []
    for ys, ls in zip(y_std, lag_std):
        if ys > 0 and ls > 0:
            colors.append('#e74c3c')  # 茅芦聵-茅芦聵 (HH)
        elif ys < 0 and ls < 0:
            colors.append('#3498db')  # 盲陆聨-盲陆聨 (LL)
        elif ys > 0 and ls < 0:
            colors.append('#f39c12')  # 茅芦聵-盲陆聨 (HL)
        else:
            colors.append('#9b59b6')  # 盲陆聨-茅芦聵 (LH)
    
    ax.scatter(y_std, lag_std, c=colors, alpha=0.7, s=50, edgecolors='white', linewidths=0.5)
    
    # 氓聫聜猫聙聝莽潞驴
    ax.axhline(0, color='gray', linewidth=0.8, linestyle='--')
    ax.axvline(0, color='gray', linewidth=0.8, linestyle='--')
    
    # 氓聸聻氓陆聮莽潞驴
    b, a = np.polyfit(y_std, lag_std, 1)
    x_line = np.linspace(y_std.min(), y_std.max(), 100)
    ax.plot(x_line, a + b * x_line, 'r--', linewidth=2, label=f"slope={b:.3f}")
    
    period_name = '莽聶陆氓陇漏' if period == 'day' else '氓陇聹茅聴麓'
    ax.set_xlabel(f'{period_name}氓聫炉猫戮戮忙聙搂茂录聢忙聽聡氓聡聠氓聦聳茂录聣', fontsize=11)
    ax.set_ylabel('莽漏潞茅聴麓忙禄聻氓聬聨', fontsize=11)
    ax.set_title(f"Moran 忙聲拢莽聜鹿氓聸戮 芒聙聰 {period_name}忙聴露忙庐碌", fontsize=13, fontweight='bold')
    
    # 氓聸戮盲戮聥
    from matplotlib.patches import Patch
    legend_elements = [
        Patch(facecolor='#e74c3c', label='HH (茅芦聵-茅芦聵)'),
        Patch(facecolor='#3498db', label='LL (盲陆聨-盲陆聨)'),
        Patch(facecolor='#f39c12', label='HL (茅芦聵-盲陆聨)'),
        Patch(facecolor='#9b59b6', label='LH (盲陆聨-茅芦聵)'),
    ]
    ax.legend(handles=legend_elements, loc='upper left', fontsize=9)

plt.tight_layout()
plt.savefig(os.path.join(BASE_DIR, '02_moran_scatter.png'), dpi=150, bbox_inches='tight', facecolor='white')
plt.show()
print("氓聸戮猫隆篓氓路虏盲驴聺氓颅聵: 02_moran_scatter.png")
```

```python
# ============================================================================
# LISA 猫聛職莽卤禄氓聢聠忙聻聬
# ============================================================================

print("\n" + "=" * 60)
print("氓卤聙茅聝篓莽漏潞茅聴麓猫聛職莽卤禄忙拢聙茅陋聦 (LISA Cluster)")
print("=" * 60)

lisa_results = {}

for period in ['day', 'night']:
    col = f'A_i_2sfca_norm_{period}'
    y = acc_gdf[col].fillna(acc_gdf[col].mean()).values
    
    lisa = Moran_Local(y, w_use, permutations=999)
    
    acc_gdf[f'lisa_q_{period}'] = lisa.q           # 猫卤隆茅聶聬莽录聳氓聫路 1-4
    acc_gdf[f'lisa_p_{period}'] = lisa.p_sim       # p氓聙录
    acc_gdf[f'lisa_z_{period}'] = lisa.z_sim       # z氓聢聠忙聲掳
    
    # 忙聵戮猫聭聴忙聙搂莽颅聸茅聙聣茂录聢p < 0.05茂录聣
    sig_mask = lisa.p_sim < 0.05
    hh = sig_mask & (lisa.q == 1)
    ll = sig_mask & (lisa.q == 3)
    hl = sig_mask & (lisa.q == 2)
    lh = sig_mask & (lisa.q == 4)
    
    period_name = '莽聶陆氓陇漏' if period == 'day' else '氓陇聹茅聴麓'
    print(f"\n{period_name}忙聴露忙庐碌 LISA 猫聛職莽卤禄:")
    print(f"  茅芦聵-茅芦聵 (HH) 氓聫炉猫戮戮忙聙搂氓炉聦猫拢聲莽聝颅莽聜鹿: {hh.sum()} 盲赂陋 (p<0.05)")
    print(f"  盲陆聨-盲陆聨 (LL) 氓聫炉猫戮戮忙聙搂猫麓芦氓聸掳氓聠路莽聜鹿: {ll.sum()} 盲赂陋 (p<0.05)")
    print(f"  茅芦聵-盲陆聨 (HL) 莽娄禄莽戮陇茅芦聵氓聙录:      {hl.sum()} 盲赂陋")
    print(f"  盲陆聨-茅芦聵 (LH) 莽娄禄莽戮陇盲陆聨氓聙录:      {lh.sum()} 盲赂陋")
    
    # LISA 氓聢聠莽卤禄忙聽聡莽颅戮
    acc_gdf[f'lisa_cluster_{period}'] = 'ns'  # not significant
    acc_gdf.loc[hh, f'lisa_cluster_{period}'] = 'HH'
    acc_gdf.loc[ll, f'lisa_cluster_{period}'] = 'LL'
    acc_gdf.loc[hl, f'lisa_cluster_{period}'] = 'HL'
    acc_gdf.loc[lh, f'lisa_cluster_{period}'] = 'LH'
    
    lisa_results[period] = {
        'HH': int(hh.sum()),
        'LL': int(ll.sum()),
        'HL': int(hl.sum()),
        'LH': int(lh.sum())
    }

print("\n芒聹聯 LISA 氓聢聠忙聻聬氓庐聦忙聢聬")
```

- 莽聶陆氓陇漏/氓陇聹茅聴麓氓聫炉猫戮戮忙聙搂

- 莽陇戮氓聦潞氓聼潞忙聹卢盲驴隆忙聛炉

```python
# ============================================================================
# Folium 盲潞陇盲潞聮氓聹掳氓聸戮 芒聙聰 LISA 猫聛職莽卤禄氓聹掳氓聸戮
# ============================================================================

import folium
from folium import plugins

# 猫庐隆莽庐聴莽聽聰莽漏露氓聦潞盲赂颅氓驴聝
center_lat = acc_gdf.geometry.centroid.y.mean()
center_lng = acc_gdf.geometry.centroid.x.mean()

# 氓聢聸氓禄潞氓潞聲氓聸戮
m = folium.Map(
    location=[center_lat, center_lng],
    zoom_start=13,
    tiles='CartoDB positron'  # 氓鹿虏氓聡聙莽職聞氓潞聲氓聸戮茂录聦茅聙聜氓聬聢忙聲掳忙聧庐氓聫炉猫搂聠氓聦聳
)

# LISA 猫聛職莽卤禄茅垄聹猫聣虏
LISA_COLORS = {
    'HH': '#e74c3c',  # 茅芦聵-茅芦聵 莽聝颅莽聜鹿
    'LL': '#3498db',  # 盲陆聨-盲陆聨 氓聠路莽聜鹿
    'HL': '#f39c12',  # 茅芦聵-盲陆聨
    'LH': '#9b59b6',  # 盲陆聨-茅芦聵
    'ns': '#cccccc'   # 盲赂聧忙聵戮猫聭聴
}

def get_popup_html(row, period='day'):
    """莽聰聼忙聢聬氓掳聫氓聦潞氓录鹿莽陋聴 HTML"""
    acc_day = row.get(f'A_i_2sfca_norm_day', 'N/A')
    acc_night = row.get(f'A_i_2sfca_norm_night', 'N/A')
    tpi = row.get('TPI', 'N/A')
    gaussian = row.get('A_i_gaussian_norm', 'N/A')
    ctype_cn = {
        'urban_village': '氓聼聨盲赂颅忙聺聭',
        'affordable_housing': '盲驴聺茅職聹忙聢驴',
        'commodity_housing': '氓聲聠氓聯聛忙聢驴',
        'high_end': '茅芦聵莽芦炉莽陇戮氓聦潞'
    }.get(row.get('community_type', ''), row.get('community_type', ''))
    
    if isinstance(acc_day, float):
        acc_day = f"{acc_day:.3f}"
    if isinstance(acc_night, float):
        acc_night = f"{acc_night:.3f}"
    if isinstance(tpi, float):
        tpi = f"{tpi:.1f}%"
    if isinstance(gaussian, float):
        gaussian = f"{gaussian:.3f}"
    
    html = f"""
    <div style='font-family:Arial,sans-serif;width:220px'>
        <h4 style='margin:0 0 8px;color:#2c3e50'>{row.get('name', 'N/A')}</h4>
        <table style='width:100%;font-size:12px;border-collapse:collapse'>
            <tr><td style='padding:3px;color:#7f8c8d'>莽卤禄氓聻聥</td><td><b>{ctype_cn}</b></td></tr>
            <tr style='background:#f8f9fa'><td style='padding:3px;color:#7f8c8d'>盲潞潞氓聫拢</td><td>{row.get('population', 'N/A')}</td></tr>
            <tr><td style='padding:3px;color:#7f8c8d'>莽聶陆氓陇漏氓聫炉猫戮戮忙聙搂</td><td style='color:#27ae60'><b>{acc_day}</b></td></tr>
            <tr style='background:#f8f9fa'><td style='padding:3px;color:#7f8c8d'>氓陇聹茅聴麓氓聫炉猫戮戮忙聙搂</td><td style='color:#e74c3c'><b>{acc_night}</b></td></tr>
            <tr><td style='padding:3px;color:#7f8c8d'>Gaussian 2SFCA</td><td><b>{gaussian}</b></td></tr>
            <tr style='background:#fff3cd'><td style='padding:3px;color:#856404'>TPI 猫麓芦氓聸掳忙聦聡忙聲掳</td><td style='color:#d35400'><b>{tpi}</b></td></tr>
            <tr><td style='padding:3px;color:#7f8c8d'>LISA 莽卤禄氓聻聥</td><td>{row.get(f'lisa_cluster_day', 'N/A')}</td></tr>
        </table>
    </div>
    """
    return folium.IFrame(html, width=240, height=200)


# 忙路禄氓聤聽 LISA 猫聛職莽卤禄氓聸戮氓卤聜
lisa_cluster_group = folium.FeatureGroup(name='LISA 猫聛職莽卤禄氓聹掳氓聸戮 (莽聶陆氓陇漏)')

for _, row in acc_gdf.iterrows():
    geom = row.geometry
    lisa_type = row.get(f'lisa_cluster_day', 'ns')
    color = LISA_COLORS.get(lisa_type, '#cccccc')
    
    # 莽聜鹿茅垄聹猫聣虏
    if geom.geom_type == 'Polygon':
        centroid = geom.centroid
    else:
        centroid = geom
    
    popup_html = get_popup_html(row)
    
    folium.CircleMarker(
        location=[centroid.y, centroid.x],
        radius=8 if lisa_type != 'ns' else 5,
        color=color,
        fill=True,
        fill_color=color,
        fill_opacity=0.8 if lisa_type != 'ns' else 0.3,
        popup=folium.Popup(popup_html, max_width=260),
        tooltip=f"{row.get('name', '')} | {lisa_type}"
    ).add_to(lisa_cluster_group)

lisa_cluster_group.add_to(m)

# 忙路禄氓聤聽氓聸戮盲戮聥
legend_html = '''
<div style='position:fixed;bottom:30px;left:30px;background:white;
    border:2px solid gray;z-index:9999;padding:10px;border-radius:6px;
    font-family:Arial,sans-serif;font-size:12px'>
<b>LISA 猫聛職莽卤禄莽卤禄氓聻聥</b><br>
<i style='background:#e74c3c;width:12px;height:12px;display:inline-block;border-radius:50%;margin-right:5px'></i> HH 茅芦聵-茅芦聵莽聝颅莽聜鹿<br>
<i style='background:#3498db;width:12px;height:12px;display:inline-block;border-radius:50%;margin-right:5px'></i> LL 盲陆聨-盲陆聨氓聠路莽聜鹿<br>
<i style='background:#f39c12;width:12px;height:12px;display:inline-block;border-radius:50%;margin-right:5px'></i> HL 茅芦聵-盲陆聨莽娄禄莽戮陇<br>
<i style='background:#9b59b6;width:12px;height:12px;display:inline-block;border-radius:50%;margin-right:5px'></i> LH 盲陆聨-茅芦聵莽娄禄莽戮陇<br>
<i style='background:#cccccc;width:12px;height:12px;display:inline-block;border-radius:50%;margin-right:5px'></i> NS 盲赂聧忙聵戮猫聭聴
</div>
'''
m.get_root().html.add_child(folium.Element(legend_html))

# 氓聸戮氓卤聜忙聨搂氓聢露
folium.LayerControl().add_to(m)

# 氓聟篓氓卤聫忙聦聣茅聮庐
plugins.Fullscreen().add_to(m)

m.save(os.path.join(BASE_DIR, '03_lisa_cluster_map.html'))
print("芒聹聯 LISA 盲潞陇盲潞聮氓聹掳氓聸戮氓路虏盲驴聺氓颅聵: 03_lisa_cluster_map.html")
display(m)
```
